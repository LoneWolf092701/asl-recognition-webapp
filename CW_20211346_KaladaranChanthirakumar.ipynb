## Part A: Application Area Review

Sign language recognition represents a critical intersection of artificial intelligence, computer vision, and accessibility technology, which address communication barriers faced by approximately 70 million deaf individuals worldwide who use sign language as their primary communication method (World Health Organization, 2024). The domain encompasses the automatic recognition and translation of visual-gestural languages. This requires sophisticated understanding of hand movements, facial expressions, body posture, and temporal dynamics that collectively convey linguistic meaning. This review examines the technological evolution, notable AI techniques, key applications to provide a comprehensive overview.

### Evolution of AI Applications in Sign Language Recognition

#### Early Phase (2015–2017)
This period relied heavily on traditional machine learning approaches, such as Support Vector Machines (SVMs), Hidden Markov Models (HMMs), and handcrafted features like Histogram of Oriented Gradients (HOG) and Scale-Invariant Feature Transform (SIFT). These methods required extensive feature engineering and struggled with the complexity and variability of sign languages, achieving accuracies around 80–90% for isolated signs (Adeyanju et al., 2021).

#### Deep Learning Era (2018–2020)
The introduction of Convolutional Neural Networks (CNNs) revolutionized the field by enabling automatic feature extraction from visual data, such as RGB and depth images. Architectures like ResNet and MobileNet achieved significant success in isolated sign recognition, with accuracies up to 99% (Aly & Aly, 2020). Long Short-Term Memory (LSTM) networks and Gated Recurrent Units (GRUs) addressed the temporal sequencing challenges inherent in continuous sign language, capturing dynamic transitions and movement epenthesis (Liao et al., 2019). Multi-modal approaches integrating RGB, depth, and skeletal data improved accuracy by 10–15% over single-modal systems (Li et al., 2020).

#### Transformer Era (2021–2025)
The adoption of transformer architectures, leveraging attention mechanisms, marked a significant leap in handling long-range dependencies and contextual understanding. Models like Sign Language Transformers achieved state-of-the-art results in end-to-end recognition and translation tasks, with accuracies exceeding 85% for continuous sign language recognition (Camgoz et al., 2020). Industry adoption surged, with Google's MediaPipe framework achieving 95.1% accuracy for micro-gestures and Apple's Vision Pro demonstrating sub-10ms latency for real-time hand tracking (Lugaresi et al., 2019; Damdoo et al., 2025).

### Notable AI Techniques and Significant Breakthroughs

**Convolutional Neural Networks (CNNs):** CNNs form the foundation for spatial feature extraction, with 2D CNNs achieving 87–96% accuracy for static signs and 3D CNNs handling spatiotemporal dynamics with 84–95% accuracy. Advanced architectures like ResNet152 and YOLOv8 have reached 99.98% accuracy for American Sign Language (ASL) alphabet recognition (Sahoo et al., 2022).

**Long Short-Term Memory (LSTM) Networks:** LSTMs excel in modeling temporal dependencies, achieving 87–99% accuracy for sequential gesture data. Hybrid CNN-LSTM models combine spatial and temporal modeling, yielding 91–95% accuracy across diverse applications (Ilham et al., 2023).

**Transformer-Based Approaches:** Transformers leverage self-attention and cross-attention mechanisms to integrate multiple modalities, improving accuracy by 2–4% over CNN-LSTM models while reducing computational requirements for edge devices (Camgoz et al., 2020; Damdoo et al., 2025).

**MediaPipe Framework:** Google's MediaPipe provides real-time hand tracking with 21-point landmark detection, achieving 90–95% accuracy with 16–20ms latency, making it ideal for mobile applications (Lugaresi et al., 2019).

### Key Applications and Developments

**Sign Language Recognition:** Isolated sign recognition has reached over 95% accuracy, while continuous sign recognition achieves 85%+ with real-time capabilities. Systems support multiple languages, including ASL, British Sign Language (BSL), and Chinese Sign Language (CSL), enhancing global accessibility (Camgoz et al., 2020).

## Part B: Comparison and Evaluation of Three AI Techniques

### Technique 1: MediaPipe Hand Tracking with LSTM Networks

**Description:** This approach combines Google's MediaPipe framework for real-time hand tracking with LSTM networks for temporal sequence modeling. MediaPipe extracts 21 three-dimensional hand landmarks per hand, capturing spatial configurations, while LSTMs model temporal dependencies crucial for continuous sign language recognition.

**Data Availability:** Requires video sequences annotated with frame-level hand positions and sign labels. Datasets like WLASL and MS-ASL are widely available, though custom data collection may enhance personalization (Li et al., 2020). Preprocessing involves frame extraction, landmark detection, coordinate normalization, and sequence padding.

**Setup Time:** Minimal due to MediaPipe's pre-trained models, with prototype development achievable in days. Training on standard GPU hardware typically completes in hours (Ilham et al., 2023).

**Results Production:** Achieves real-time processing at 30fps with accuracies above 90% for isolated signs and 88.23% for continuous signs (Sahoo et al., 2022; Srivastava et al., 2024).

**Output Quality:** Provides spatial hand configurations, temporal sequence classifications, and confidence scores. Visualizations of hand trajectories enhance interpretability. Limitations include challenges with complex grammatical structures and non-manual features like facial expressions (Lugaresi et al., 2019).

**Strengths and Weaknesses:** Excels in real-time deployment and robustness across lighting conditions but struggles with non-manual markers and long-range dependencies.

### Technique 2: Transformer-Based Multimodal Fusion

**Description:** Transformer architectures integrate multiple modalities (hand movements, facial expressions, body poses) using self-attention and cross-attention mechanisms to model long-range dependencies and contextual relationships, ideal for continuous sign language recognition.

**Data Availability:** Requires comprehensive multimodal datasets like How2Sign and DGS Corpus, which synchronize hand tracking, facial landmarks, and pose estimation. High-quality annotations for non-manual markers are essential but complex to obtain (Camgoz et al., 2020).

**Setup Time:** Longer due to the complexity of multimodal feature extraction and transformer implementation. Development typically takes weeks, with training requiring days to weeks on high-end GPUs (Zhang et al., 2024).

**Results Production:** Achieves state-of-the-art accuracy (e.g., 85% on complex sentence structures) but may require optimization for real-time performance. Offline processing yields superior results (Camgoz et al., 2020).

**Output Quality:** Includes sign classification, confidence distributions, attention visualizations, and potential for end-to-end translation. Attention weights reveal modality contributions, enhancing interpretability.

**Strengths and Weaknesses:** Excels in handling linguistic complexity but is computationally intensive, limiting edge device deployment without optimization (e.g., SignEdgeLVM reduces memory consumption by 99.93%) (Zhang et al., 2024).

### Technique 3: 3D Convolutional Neural Networks with Spatial-Temporal Processing

**Description:** 3D CNNs treat video sequences as volumetric data, applying convolution operations across spatial and temporal dimensions to capture dynamic gestures without explicit feature engineering.

**Data Availability:** Requires high-quality video datasets with temporal annotations, such as RWTH-PHOENIX14T and CSL. Preprocessing includes video segmentation, frame standardization, and data augmentation (Huang et al., 2015).

**Setup Time:** Varies from days for simpler architectures to weeks for complex models with residual connections or attention mechanisms. Training demands multiple GPUs and extends from days to weeks (Liao et al., 2019).

**Results Production:** Typically suited for offline processing due to computational intensity, though optimized versions achieve near real-time performance. Accuracies reach 93.9% on datasets like LSA64 (Huang et al., 2018).

**Output Quality:** Provides hierarchical feature representations, classification confidence scores, and temporal localization. Excels in capturing subtle movement patterns but offers less interpretability than transformers.

**Strengths and Weaknesses:** Robust for dynamic gestures but computationally expensive and less effective for non-manual features compared to multimodal transformers.

### Technique Comparison Summary

| Technique | Accuracy Range | Strengths | Weaknesses |
|-----------|----------------|-----------|------------|
| MediaPipe + LSTM | 88–99% | Real-time processing, low setup time | Limited handling of non-manual features |
| Transformers | 74–85% | Contextual understanding, multimodal integration | High computational cost, complex setup |
| 3D CNNs | 84–93.9% | Spatiotemporal feature capture, no feature engineering | Computationally intensive, less interpretable |

````python
# Cell 5: MediaPipe Hand Landmark Extractor
"""
Implementing MediaPipe-based hand landmark extraction system.

MediaPipe detects 21 landmarks per hand, providing precise spatial coordinates
that capture hand shape and finger positions essential for sign language recognition.
"""

class HandLandmarkExtractor:
    """
    MediaPipe-based hand landmark extraction with comprehensive preprocessing.
    
    This class handles:
    - Hand detection and landmark extraction
    - Coordinate normalization for translation invariance
    - Visualization of detected landmarks
    - Robust handling of missing or unclear hands
    """
    
    def __init__(self, static_image_mode=True, max_num_hands=1, 
                 min_detection_confidence=0.5, min_tracking_confidence=0.5):
        """
        Initialize MediaPipe hands solution with optimized parameters.
        
        Args:
            static_image_mode: True for single images, False for video streams
            max_num_hands: Maximum number of hands to detect (1 for ASL alphabet)
            min_detection_confidence: Minimum confidence for hand detection
            min_tracking_confidence: Minimum confidence for hand tracking
        """
        self.mp_hands = mp.solutions.hands
        self.hands = self.mp_hands.Hands(
            static_image_mode=static_image_mode,
            max_num_hands=max_num_hands,
            min_detection_confidence=min_detection_confidence,
            min_tracking_confidence=min_tracking_confidence
        )
        self.mp_drawing = mp.solutions.drawing_utils
        
        # Track processing statistics
        self.processed_images = 0
        self.successful_detections = 0
        
    def extract_landmarks(self, image):
        """
        Extract hand landmarks from a single image.
        
        Args:
            image: Input image (PIL Image or numpy array)
            
        Returns:
            landmarks: Normalized landmark coordinates (42 features for 1 hand)
            success: Boolean indicating if hand was detected
            annotated_image: Image with drawn landmarks (for visualization)
        """
        # Convert PIL Image to numpy array if necessary
        if isinstance(image, Image.Image):
            image_array = np.array(image)
        else:
            image_array = image
        
        # Ensure image is in RGB format for MediaPipe
        if len(image_array.shape) == 3:
            if image_array.shape[2] == 4:  # RGBA
                rgb_image = cv2.cvtColor(image_array, cv2.COLOR_RGBA2RGB)
            elif image_array.shape[2] == 3:  # Already RGB or BGR
                # MediaPipe expects RGB, OpenCV loads as BGR
                rgb_image = cv2.cvtColor(image_array, cv2.COLOR_BGR2RGB)
            else:
                rgb_image = image_array
        else:
            # Grayscale image
            rgb_image = cv2.cvtColor(image_array, cv2.COLOR_GRAY2RGB)
        
        # Process the image with MediaPipe
        results = self.hands.process(rgb_image)
        
        # Initialize landmark array (21 landmarks × 2 coordinates = 42 features)
        landmarks = np.zeros(42)
        success = False
        
        # Create annotated image for visualization
        annotated_image = rgb_image.copy()
        
        if results.multi_hand_landmarks:
            # Extract landmarks from the first detected hand
            hand_landmarks = results.multi_hand_landmarks[0]
            
            # Extract x, y coordinates for each landmark
            for i, landmark in enumerate(hand_landmarks.landmark):
                landmarks[i * 2] = landmark.x      # x coordinate
                landmarks[i * 2 + 1] = landmark.y  # y coordinate
            
            success = True
            self.successful_detections += 1
            
            # Draw landmarks on image for visualization
            self.mp_drawing.draw_landmarks(
                annotated_image,
                hand_landmarks,
                self.mp_hands.HAND_CONNECTIONS
            )
        
        self.processed_images += 1
        return landmarks, success, annotated_image
    
    def normalize_landmarks(self, landmarks):
        """
        Normalize landmarks relative to wrist position for translation invariance.
        
        Args:
            landmarks: Raw landmark coordinates
            
        Returns:
            normalized_landmarks: Normalized coordinates relative to wrist
        """
        if np.all(landmarks == 0):
            return landmarks
        
        normalized = landmarks.copy()
        
        # Get wrist position (landmark 0)
        wrist_x = landmarks[0]
        wrist_y = landmarks[1]
        
        # Normalize all landmarks relative to wrist position
        for i in range(21):
            normalized[i * 2] = landmarks[i * 2] - wrist_x        # x coordinate
            normalized[i * 2 + 1] = landmarks[i * 2 + 1] - wrist_y  # y coordinate
        
        return normalized
    
    def get_statistics(self):
        """
        Get processing statistics for monitoring extraction performance.
        
        Returns:
            dict: Processing statistics including success rate
        """
        success_rate = (self.successful_detections / self.processed_images * 100) if self.processed_images > 0 else 0
        
        return {
            'processed_images': self.processed_images,
            'successful_detections': self.successful_detections,
            'success_rate': success_rate,
            'failed_detections': self.processed_images - self.successful_detections
        }

# Initialize the hand landmark extractor
print("Initializing MediaPipe Hand Landmark Extractor...")
landmark_extractor = HandLandmarkExtractor(
    static_image_mode=True,
    max_num_hands=1,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)
print("Hand landmark extractor initialized successfully!")

# Test the extractor with a sample image
print("\nTesting landmark extraction with sample image...")
try:
    # Load a sample image from the dataset
    sample_class = 'A'
    sample_class_path = os.path.join(TRAIN_PATH, sample_class)
    
    if os.path.exists(sample_class_path):
        sample_images = [f for f in os.listdir(sample_class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        if sample_images:
            # Try multiple sample images to find one that works
            test_success = False
            for i, sample_image_file in enumerate(sample_images[:5]):  # Try first 5 images
                sample_image_path = os.path.join(sample_class_path, sample_image_file)
                
                try:
                    # Load and process the sample image
                    sample_image = Image.open(sample_image_path)
                    
                    # Convert to RGB if needed
                    if sample_image.mode != 'RGB':
                        sample_image = sample_image.convert('RGB')
                    
                    landmarks, success, annotated_image = landmark_extractor.extract_landmarks(sample_image)
                    
                    if success:
                        print(f"Successfully extracted {len(landmarks)} landmark features from image {i+1}")
                        print(f"Sample landmarks (first 10): {landmarks[:10]}")
                        
                        # Visualize the extraction
                        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
                        
                        axes[0].imshow(sample_image)
                        axes[0].set_title('Original Image', fontsize=14, fontweight='bold')
                        axes[0].axis('off')
                        
                        axes[1].imshow(annotated_image)
                        axes[1].set_title('Detected Landmarks', fontsize=14, fontweight='bold')
                        axes[1].axis('off')
                        
                        plt.tight_layout()
                        plt.show()
                        
                        test_success = True
                        break
                        
                except Exception as e:
                    print(f"Error with image {i+1}: {e}")
                    continue
            
            if not test_success:
                print("Could not extract landmarks from any of the first 5 test images")
                print("This may indicate issues with image format or MediaPipe configuration")
        else:
            print(f"No image files found in {sample_class_path}")
    else:
        print(f"Sample class '{sample_class}' not found in dataset")
        
except Exception as e:
    print(f"Error testing landmark extraction: {e}")
````

````python
# Cell 6: Dataset Processing and Feature Extraction
"""
Processing the entire ASL dataset to extract MediaPipe landmarks from all images.
"""

def process_asl_dataset(train_path, output_path, max_images_per_class=500):
    """
    Process the entire ASL dataset to extract hand landmarks.
    
    Args:
        train_path: Path to the training data
        output_path: Path to save processed features
        max_images_per_class: Maximum images to process per class (for memory management)
    
    Returns:
        dict: Processing statistics and results
    """
    # Initialize multiple extractors for different detection strategies
    primary_extractor = HandLandmarkExtractor(
        static_image_mode=True,
        max_num_hands=1,
        min_detection_confidence=0.4,
        min_tracking_confidence=0.3
    )
    
    backup_extractor = HandLandmarkExtractor(
        static_image_mode=True,
        max_num_hands=1,
        min_detection_confidence=0.25,
        min_tracking_confidence=0.2
    )
    
    # Prepare data storage
    features = []
    labels = []
    failed_images = []
    detailed_failures = []
    
    # Get all class directories
    class_dirs = [d for d in os.listdir(train_path) if os.path.isdir(os.path.join(train_path, d))]
    
    # Filter out non-alphabet classes for this implementation
    alphabet_classes = [d for d in class_dirs if len(d) == 1 and d.isalpha()]
    alphabet_classes.sort()
    
    print(f"Processing {len(alphabet_classes)} alphabet classes...")
    print(f"Classes: {alphabet_classes}")
    
    # Process each class
    for class_idx, class_name in enumerate(alphabet_classes):
        print(f"\nProcessing class {class_idx + 1}/{len(alphabet_classes)}: {class_name}")
        
        class_path = os.path.join(train_path, class_name)
        image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        # Limit number of images per class to manage memory
        if len(image_files) > max_images_per_class:
            image_files = image_files[:max_images_per_class]
        
        class_features = []
        class_labels = []
        class_failures = 0
        
        # Process images with progress bar
        for image_file in tqdm(image_files, desc=f"Class {class_name}", leave=False):
            image_path = os.path.join(class_path, image_file)
            
            try:
                # Load and validate image
                image = Image.open(image_path)
                
                # Convert to RGB if needed
                if image.mode != 'RGB':
                    image = image.convert('RGB')
                
                # Get image dimensions
                width, height = image.size
                
                # Skip images that are too small or too large
                if width < 50 or height < 50:
                    detailed_failures.append(f"{image_path}: Too small {width}x{height}")
                    failed_images.append(image_path)
                    class_failures += 1
                    continue
                
                # Preprocessing for better detection
                img_array = np.array(image)
                
                # Apply contrast adjustment
                img_array = cv2.convertScaleAbs(img_array, alpha=1.1, beta=5)
                
                # Resize to optimal range for MediaPipe
                if width > 800 or height > 800:
                    if width > height:
                        new_width = 800
                        new_height = int(height * (800 / width))
                    else:
                        new_height = 800
                        new_width = int(width * (800 / height))
                    img_array = cv2.resize(img_array, (new_width, new_height), interpolation=cv2.INTER_AREA)
                elif width < 200 or height < 200:
                    scale_factor = max(200 / width, 200 / height)
                    new_width = int(width * scale_factor)
                    new_height = int(height * scale_factor)
                    img_array = cv2.resize(img_array, (new_width, new_height), interpolation=cv2.INTER_CUBIC)
                
                processed_image = Image.fromarray(img_array)
                
                # First detection attempt with primary extractor
                landmarks, success, _ = primary_extractor.extract_landmarks(processed_image)
                
                # Second attempt with backup extractor if first fails
                if not success:
                    landmarks, success, _ = backup_extractor.extract_landmarks(processed_image)
                
                # Third attempt with original image if processing failed
                if not success:
                    landmarks, success, _ = backup_extractor.extract_landmarks(image)
                
                # Fourth attempt with slight blur to reduce noise
                if not success:
                    blurred_array = cv2.GaussianBlur(img_array, (3, 3), 0)
                    blurred_image = Image.fromarray(blurred_array)
                    landmarks, success, _ = backup_extractor.extract_landmarks(blurred_image)
                
                if success:
                    # Validate landmarks contain meaningful data
                    if np.any(landmarks != 0) and not np.any(np.isnan(landmarks)) and not np.any(np.isinf(landmarks)):
                        # Check for reasonable landmark values
                        if np.all(landmarks >= -2.0) and np.all(landmarks <= 2.0):
                            # Normalize landmarks
                            normalized_landmarks = primary_extractor.normalize_landmarks(landmarks)
                            
                            # Final validation after normalization
                            if not np.any(np.isnan(normalized_landmarks)) and not np.any(np.isinf(normalized_landmarks)):
                                class_features.append(normalized_landmarks)
                                class_labels.append(class_name)
                            else:
                                detailed_failures.append(f"{image_path}: Invalid after normalization")
                                failed_images.append(image_path)
                                class_failures += 1
                        else:
                            detailed_failures.append(f"{image_path}: Landmark values out of range")
                            failed_images.append(image_path)
                            class_failures += 1
                    else:
                        detailed_failures.append(f"{image_path}: Invalid landmark data")
                        failed_images.append(image_path)
                        class_failures += 1
                else:
                    detailed_failures.append(f"{image_path}: All detection attempts failed")
                    failed_images.append(image_path)
                    class_failures += 1
                    
            except Exception as e:
                detailed_failures.append(f"{image_path}: Exception {str(e)}")
                failed_images.append(image_path)
                class_failures += 1
        
        # Add to main collections
        features.extend(class_features)
        labels.extend(class_labels)
        
        # Report class-level statistics
        total_class_images = len(image_files)
        success_count = len(class_features)
        success_rate = (success_count / total_class_images * 100) if total_class_images > 0 else 0
        
        print(f"  Processed {success_count}/{total_class_images} images successfully ({success_rate:.1f}%)")
        if class_failures > 0:
            print(f"  Failed: {class_failures} images")
    
    # Convert to numpy arrays
    features = np.array(features) if features else np.array([]).reshape(0, 42)
    labels = np.array(labels) if labels else np.array([])
    
    # Create output directory
    os.makedirs(output_path, exist_ok=True)
    
    # Save processed data
    np.savez_compressed(
        os.path.join(output_path, 'asl_features.npz'),
        features=features,
        labels=labels
    )
    
    # Get processing statistics
    total_processed = primary_extractor.processed_images + backup_extractor.processed_images
    total_successful = primary_extractor.successful_detections + backup_extractor.successful_detections
    overall_success_rate = (total_successful / total_processed * 100) if total_processed > 0 else 0
    
    # Save processing statistics
    processing_stats = {
        'total_classes': len(alphabet_classes),
        'total_images_processed': len(features),
        'failed_images': len(failed_images),
        'success_rate': overall_success_rate,
        'feature_shape': features.shape,
        'classes': alphabet_classes
    }
    
    # Save detailed statistics
    with open(os.path.join(output_path, 'processing_stats.txt'), 'w') as f:
        for key, value in processing_stats.items():
            f.write(f"{key}: {value}\n")
    
    # Save failure analysis for debugging
    if detailed_failures:
        with open(os.path.join(output_path, 'failure_analysis.txt'), 'w') as f:
            f.write("FAILURE ANALYSIS\n")
            f.write("================\n\n")
            
            # Count failure types
            failure_types = {}
            for failure in detailed_failures:
                if "All detection attempts failed" in failure:
                    failure_types['detection_failed'] = failure_types.get('detection_failed', 0) + 1
                elif "Too small" in failure:
                    failure_types['too_small'] = failure_types.get('too_small', 0) + 1
                elif "Invalid" in failure:
                    failure_types['invalid_data'] = failure_types.get('invalid_data', 0) + 1
                elif "out of range" in failure:
                    failure_types['out_of_range'] = failure_types.get('out_of_range', 0) + 1
                elif "Exception" in failure:
                    failure_types['exceptions'] = failure_types.get('exceptions', 0) + 1
                else:
                    failure_types['other'] = failure_types.get('other', 0) + 1
            
            f.write("Failure Type Summary:\n")
            for failure_type, count in failure_types.items():
                f.write(f"  {failure_type}: {count}\n")
            
            f.write(f"\nDetailed Failures (first 100):\n")
            for failure in detailed_failures[:100]:
                f.write(f"{failure}\n")
    
    # Provide processing feedback
    final_success_rate = overall_success_rate
    print(f"\nProcessing completed with {final_success_rate:.1f}% overall success rate")
    
    if final_success_rate < 20:
        print("Consider reviewing dataset quality and detection parameters")
    elif final_success_rate < 40:
        print("Moderate success rate - may need additional preprocessing")
    else:
        print("Good success rate for training data")
    
    return processing_stats, features, labels

# Process the dataset
print("PROCESSING ASL DATASET")
print("="*50)

# Verify dataset path exists
if not os.path.exists(TRAIN_PATH):
    print(f"ERROR: Training path not found: {TRAIN_PATH}")
    print("Please ensure the ASL dataset is properly extracted")
else:
    # Create output directory for processed data
    output_path = os.path.join(BASE_DIR, 'data', 'processed')
    
    # Process with multiple detection strategies
    processing_stats, features, labels = process_asl_dataset(TRAIN_PATH, output_path, max_images_per_class=300)
    
    print(f"\nDATASET PROCESSING COMPLETED!")
    print(f"Processing Statistics:")
    print(f"  - Total classes: {processing_stats['total_classes']}")
    print(f"  - Images processed: {processing_stats['total_images_processed']:,}")
    print(f"  - Failed extractions: {processing_stats['failed_images']:,}")
    print(f"  - Success rate: {processing_stats['success_rate']:.1f}%")
    print(f"  - Feature shape: {processing_stats['feature_shape']}")
    print(f"  - Classes: {processing_stats['classes']}")
    
    # Additional validation
    if len(features) > 0:
        print(f"\nFeature validation:")
        print(f"  - Feature range: [{np.min(features):.3f}, {np.max(features):.3f}]")
        print(f"  - NaN values: {np.sum(np.isnan(features))}")
        print(f"  - Infinite values: {np.sum(np.isinf(features))}")
        print(f"  - All-zero samples: {np.sum(np.all(features == 0, axis=1))}")
    else:
        print("\nERROR: No features extracted! Check dataset and processing parameters.")
````

````python
# Different Cell 6 (Testing phase) - Might take more time as it uses different extractorssss
# Cell 6: Improved Dataset Processing and Feature Extraction
"""
Dataset processing pipeline with much better landmark extraction success rates.
"""

def process_asl_dataset_v2(train_path, output_path, max_images_per_class=500):
    """
    Improved ASL dataset processing with enhanced landmark extraction strategies.
    
    Args:
        train_path: Path to the training data
        output_path: Path to save processed features
        max_images_per_class: Maximum images to process per class
    
    Returns:
        dict: Processing statistics and results
    """
    
    # Initialize multiple extractors with progressive sensitivity levels
    # Primary extractor - standard settings
    primary_extractor = HandLandmarkExtractor(
        static_image_mode=True,
        max_num_hands=1,
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5
    )
    
    # Secondary extractor - more sensitive
    secondary_extractor = HandLandmarkExtractor(
        static_image_mode=True,
        max_num_hands=1,
        min_detection_confidence=0.3,
        min_tracking_confidence=0.3
    )
    
    # Tertiary extractor - highly sensitive for difficult cases
    tertiary_extractor = HandLandmarkExtractor(
        static_image_mode=True,
        max_num_hands=1,
        min_detection_confidence=0.1,
        min_tracking_confidence=0.1
    )
    
    def preprocess_image_for_detection(image):
        """
        Advanced image preprocessing to improve hand detection rates.
        
        Args:
            image: PIL Image
            
        Returns:
            List of processed image variants
        """
        variants = []
        
        # Convert to numpy array for processing
        img_array = np.array(image)
        
        # Original image
        variants.append(image)
        
        # Contrast and brightness adjustments
        # High contrast version
        high_contrast = np.clip(img_array * 1.5 + 20, 0, 255).astype(np.uint8)
        variants.append(Image.fromarray(high_contrast))
        
        # Low contrast version (sometimes helps with overexposed images)
        low_contrast = np.clip(img_array * 0.8 + 30, 0, 255).astype(np.uint8)
        variants.append(Image.fromarray(low_contrast))
        
        # Histogram equalization for better contrast distribution
        if len(img_array.shape) == 3:
            # Convert to YUV for better histogram equalization
            yuv = cv2.cvtColor(img_array, cv2.COLOR_RGB2YUV)
            yuv[:,:,0] = cv2.equalizeHist(yuv[:,:,0])
            hist_eq = cv2.cvtColor(yuv, cv2.COLOR_YUV2RGB)
            variants.append(Image.fromarray(hist_eq))
        
        # Gaussian blur to reduce noise (can help with detection)
        blurred = cv2.GaussianBlur(img_array, (3, 3), 0)
        variants.append(Image.fromarray(blurred))
        
        # Sharpening filter
        kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
        sharpened = cv2.filter2D(img_array, -1, kernel)
        sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)
        variants.append(Image.fromarray(sharpened))
        
        return variants
    
    def try_comprehensive_extraction(image, image_path):
        """
        Comprehensive landmark extraction using multiple strategies.
        
        Args:
            image: PIL Image
            image_path: Path to image for debugging
            
        Returns:
            Tuple of (landmarks, success, method_used)
        """
        extractors = [
            (primary_extractor, "primary"),
            (secondary_extractor, "secondary"), 
            (tertiary_extractor, "tertiary")
        ]
        
        # Try with original image first
        for extractor, method_name in extractors:
            try:
                landmarks, success, _ = extractor.extract_landmarks(image)
                if success and validate_landmarks(landmarks):
                    return landmarks, True, f"{method_name}_original"
            except:
                continue
        
        # Try with preprocessed variants
        try:
            variants = preprocess_image_for_detection(image)
            
            for variant_idx, variant in enumerate(variants[1:], 1):  # Skip original (already tried)
                for extractor, method_name in extractors:
                    try:
                        landmarks, success, _ = extractor.extract_landmarks(variant)
                        if success and validate_landmarks(landmarks):
                            return landmarks, True, f"{method_name}_variant_{variant_idx}"
                    except:
                        continue
        except:
            pass
        
        return None, False, "all_failed"
    
    def validate_landmarks(landmarks):
        """
        Validate landmark quality to ensure good training data.
        
        Args:
            landmarks: Extracted landmark coordinates
            
        Returns:
            Boolean indicating if landmarks are valid
        """
        if landmarks is None or len(landmarks) != 42:
            return False
        
        # Check for valid numeric values
        if np.any(np.isnan(landmarks)) or np.any(np.isinf(landmarks)):
            return False
        
        # Check if landmarks are within reasonable bounds
        if not np.all(landmarks >= -2.0) or not np.all(landmarks <= 2.0):
            return False
        
        # Check if there's actual hand movement (not all zeros)
        if np.all(landmarks == 0):
            return False
        
        # Check for reasonable landmark spread (hand should occupy some space)
        x_coords = landmarks[::2]  # Every other element is x-coordinate
        y_coords = landmarks[1::2]  # Every other element is y-coordinate
        
        x_range = np.max(x_coords) - np.min(x_coords)
        y_range = np.max(y_coords) - np.min(y_coords)
        
        # Hand should occupy at least some minimum space
        if x_range < 0.05 or y_range < 0.05:
            return False
        
        return True
    
    # Prepare data storage
    features = []
    labels = []
    processing_stats = {}
    method_counts = {}
    
    # Get all class directories
    class_dirs = [d for d in os.listdir(train_path) if os.path.isdir(os.path.join(train_path, d))]
    
    # Filter out non-alphabet classes for this implementation
    alphabet_classes = [d for d in class_dirs if len(d) == 1 and d.isalpha()]
    alphabet_classes.sort()
    
    print(f"Processing {len(alphabet_classes)} alphabet classes with improved extraction...")
    print(f"Classes: {alphabet_classes}")
    
    # Process each class
    for class_idx, class_name in enumerate(alphabet_classes):
        print(f"\nProcessing class {class_idx + 1}/{len(alphabet_classes)}: {class_name}")
        
        class_path = os.path.join(train_path, class_name)
        image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        # Limit number of images per class
        if len(image_files) > max_images_per_class:
            # Randomly sample for better variety
            import random
            random.seed(42)  # For reproducibility
            image_files = random.sample(image_files, max_images_per_class)
        
        class_features = []
        class_labels = []
        class_methods = {}
        
        # Process images with progress bar
        for image_file in tqdm(image_files, desc=f"Class {class_name}", leave=False):
            image_path = os.path.join(class_path, image_file)
            
            try:
                # Load and validate image
                image = Image.open(image_path)
                
                # Convert to RGB format consistently
                if image.mode != 'RGB':
                    image = image.convert('RGB')
                
                # Check image dimensions
                width, height = image.size
                
                # Skip extremely small images
                if width < 60 or height < 60:
                    continue
                
                # Resize very large images for consistent processing
                if width > 1200 or height > 1200:
                    max_dimension = max(width, height)
                    scale_factor = 800 / max_dimension
                    new_width = int(width * scale_factor)
                    new_height = int(height * scale_factor)
                    image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
                
                # Try comprehensive extraction
                landmarks, success, method_used = try_comprehensive_extraction(image, image_path)
                
                if success:
                    # Normalize landmarks
                    normalized_landmarks = primary_extractor.normalize_landmarks(landmarks)
                    
                    # Final validation after normalization
                    if validate_landmarks(normalized_landmarks):
                        class_features.append(normalized_landmarks)
                        class_labels.append(class_name)
                        
                        # Track extraction methods for analysis
                        class_methods[method_used] = class_methods.get(method_used, 0) + 1
                        method_counts[method_used] = method_counts.get(method_used, 0) + 1
                    
            except Exception as e:
                # Skip problematic images
                continue
        
        # Add to main collections
        features.extend(class_features)
        labels.extend(class_labels)
        
        # Store class-level statistics
        total_attempted = len(image_files)
        successful = len(class_features)
        success_rate = (successful / total_attempted * 100) if total_attempted > 0 else 0
        
        processing_stats[class_name] = {
            'attempted': total_attempted,
            'successful': successful,
            'success_rate': success_rate,
            'methods': class_methods
        }
        
        print(f"  Processed {successful}/{total_attempted} images successfully ({success_rate:.1f}%)")
        if class_methods:
            print(f"  Methods used: {dict(class_methods)}")
    
    # Convert to numpy arrays
    features = np.array(features) if features else np.array([]).reshape(0, 42)
    labels = np.array(labels) if labels else np.array([])
    
    # Save processed data
    os.makedirs(output_path, exist_ok=True)
    
    np.savez_compressed(
        os.path.join(output_path, 'asl_features_improved.npz'),
        features=features,
        labels=labels
    )
    
    # Calculate overall statistics
    total_attempted = sum([stats['attempted'] for stats in processing_stats.values()])
    total_successful = len(features)
    overall_success_rate = (total_successful / total_attempted * 100) if total_attempted > 0 else 0
    
    # Save comprehensive processing statistics
    final_stats = {
        'total_classes': len(alphabet_classes),
        'total_attempted': total_attempted,
        'total_successful': total_successful,
        'overall_success_rate': overall_success_rate,
        'feature_shape': features.shape,
        'classes': alphabet_classes,
        'class_statistics': processing_stats,
        'extraction_methods': method_counts
    }
    
    # Save detailed statistics
    with open(os.path.join(output_path, 'improved_processing_stats.txt'), 'w') as f:
        f.write("IMPROVED ASL DATASET PROCESSING STATISTICS\n")
        f.write("=" * 50 + "\n\n")
        f.write(f"Overall Results:\n")
        f.write(f"  Total classes: {len(alphabet_classes)}\n")
        f.write(f"  Total images attempted: {total_attempted:,}\n")
        f.write(f"  Total images successful: {total_successful:,}\n")
        f.write(f"  Overall success rate: {overall_success_rate:.1f}%\n")
        f.write(f"  Feature shape: {features.shape}\n\n")
        
        f.write("Extraction Methods Used:\n")
        for method, count in method_counts.items():
            percentage = (count / total_successful * 100) if total_successful > 0 else 0
            f.write(f"  {method}: {count:,} ({percentage:.1f}%)\n")
        
        f.write("\nPer-Class Statistics:\n")
        for class_name in alphabet_classes:
            stats = processing_stats.get(class_name, {})
            f.write(f"\n{class_name}:\n")
            f.write(f"  Attempted: {stats.get('attempted', 0)}\n")
            f.write(f"  Successful: {stats.get('successful', 0)}\n")
            f.write(f"  Success rate: {stats.get('success_rate', 0):.1f}%\n")
            if 'methods' in stats:
                f.write(f"  Methods: {stats['methods']}\n")
    
    return final_stats, features, labels

# Run the improved dataset processing
print("IMPROVED ASL DATASET PROCESSING")
print("=" * 50)
print("This version uses:")
print("  - Multiple extraction strategies")
print("  - Advanced image preprocessing")
print("  - Progressive sensitivity levels")
print("  - Comprehensive quality validation")
print()

# Verify dataset path exists
if not os.path.exists(TRAIN_PATH):
    print(f"ERROR: Training path not found: {TRAIN_PATH}")
    print("Please ensure the ASL dataset is properly extracted")
else:
    # Create output directory for processed data
    output_path = os.path.join(BASE_DIR, 'data', 'processed')
    
    # Process with improved pipeline
    improved_stats, improved_features, improved_labels = process_asl_dataset_v2(
        TRAIN_PATH, output_path, max_images_per_class=300
    )
    
    print(f"\nIMPROVED DATASET PROCESSING COMPLETED!")
    print(f"Processing Statistics:")
    print(f"  - Total classes: {improved_stats['total_classes']}")
    print(f"  - Images attempted: {improved_stats['total_attempted']:,}")
    print(f"  - Images processed: {improved_stats['total_successful']:,}")
    print(f"  - Success rate: {improved_stats['overall_success_rate']:.1f}%")
    print(f"  - Feature shape: {improved_stats['feature_shape']}")
    
    # Show improvement comparison
    if improved_stats['overall_success_rate'] > 3.0:
        improvement = improved_stats['overall_success_rate'] - 3.0
        print(f"  - Improvement: +{improvement:.1f}% over previous processing")
    
    print(f"\nExtraction Methods Breakdown:")
    for method, count in improved_stats['extraction_methods'].items():
        percentage = (count / improved_stats['total_successful'] * 100) if improved_stats['total_successful'] > 0 else 0
        print(f"  - {method}: {count:,} images ({percentage:.1f}%)")
    
    # Show class balance analysis
    print(f"\nClass Balance Analysis:")
    class_counts = {}
    for label in improved_labels:
        class_counts[label] = class_counts.get(label, 0) + 1
    
    min_count = min(class_counts.values()) if class_counts else 0
    max_count = max(class_counts.values()) if class_counts else 0
    balance_ratio = max_count / min_count if min_count > 0 else float('inf')
    
    print(f"  - Min samples per class: {min_count}")
    print(f"  - Max samples per class: {max_count}")
    print(f"  - Balance ratio: {balance_ratio:.2f}")
    
    if balance_ratio > 3.0:
        print(f"  - Warning: Significant class imbalance detected")
    else:
        print(f"  - Good class balance achieved")
    
    # Update global variables for use in subsequent cells
    features = improved_features
    labels = improved_labels
    
    print(f"\nFeature validation:")
    if len(features) > 0:
        print(f"  - Feature range: [{np.min(features):.3f}, {np.max(features):.3f}]")
        print(f"  - NaN values: {np.sum(np.isnan(features))}")
        print(f"  - Infinite values: {np.sum(np.isinf(features))}")
        print(f"  - All-zero samples: {np.sum(np.all(features == 0, axis=1))}")
        print(f"  - Ready for training: {'Yes' if improved_stats['overall_success_rate'] > 30 else 'Needs review'}")
    else:
        print("  - ERROR: No features extracted! Check dataset and processing parameters.")
````

````python
# Cell 7: Sequence Creation for LSTM Training
"""
Creating temporal sequences from static landmark features for LSTM training.
"""

class ASLSequenceGenerator:
    """
    Generate temporal sequences from static ASL landmark features.
    
    This class creates realistic temporal sequences by:
    - Adding natural hand movement variations
    - Simulating approach and hold phases of signing
    - Providing data augmentation for robust training
    """
    
    def __init__(self, sequence_length=30, augmentation_factor=5):
        """
        Initialize the sequence generator.
        
        Args:
            sequence_length: Number of frames in each sequence
            augmentation_factor: Number of sequences to generate per static landmark
        """
        self.sequence_length = sequence_length
        self.augmentation_factor = augmentation_factor
        
    def create_sequence(self, landmarks, add_noise=True, add_movement=True):
        """
        Create a temporal sequence from static landmarks.
        
        Args:
            landmarks: Static hand landmarks (42 features)
            add_noise: Whether to add natural variation noise
            add_movement: Whether to add movement patterns
            
        Returns:
            sequence: Temporal sequence of landmarks
        """
        sequence = []
        
        for frame_idx in range(self.sequence_length):
            # Start with base landmarks
            frame_landmarks = landmarks.copy()
            
            if add_movement:
                # Add natural movement pattern (approach - hold - release)
                progress = frame_idx / self.sequence_length
                
                # Approach phase (0.0 to 0.3): slight movement toward final position
                if progress < 0.3:
                    movement_factor = 0.02 * (1 - progress / 0.3)
                    movement_noise = np.random.normal(0, movement_factor, landmarks.shape)
                    frame_landmarks += movement_noise
                
                # Hold phase (0.3 to 0.7): stable with minor tremor
                elif progress < 0.7:
                    tremor_factor = 0.005
                    tremor_noise = np.random.normal(0, tremor_factor, landmarks.shape)
                    frame_landmarks += tremor_noise
                
                # Release phase (0.7 to 1.0): slight movement away
                else:
                    movement_factor = 0.015 * (progress - 0.7) / 0.3
                    movement_noise = np.random.normal(0, movement_factor, landmarks.shape)
                    frame_landmarks += movement_noise
            
            if add_noise:
                # Add natural variation and measurement noise
                noise_factor = 0.008
                noise = np.random.normal(0, noise_factor, landmarks.shape)
                frame_landmarks += noise
            
            sequence.append(frame_landmarks)
        
        return np.array(sequence)
    
    def generate_sequences(self, features, labels):
        """
        Generate multiple sequences from static features.
        
        Args:
            features: Static landmark features
            labels: Corresponding labels
            
        Returns:
            sequences: Generated temporal sequences
            sequence_labels: Corresponding labels
        """
        sequences = []
        sequence_labels = []
        
        print(f"Generating {self.augmentation_factor} sequences per static landmark...")
        
        for i, (landmark, label) in enumerate(tqdm(zip(features, labels), total=len(features))):
            for aug_idx in range(self.augmentation_factor):
                # Create sequence with different augmentation parameters
                sequence = self.create_sequence(
                    landmark, 
                    add_noise=True, 
                    add_movement=True
                )
                
                sequences.append(sequence)
                sequence_labels.append(label)
        
        return np.array(sequences), np.array(sequence_labels)

# Create sequence generator
print("CREATING TEMPORAL SEQUENCES FOR LSTM TRAINING")
print("="*55)

sequence_generator = ASLSequenceGenerator(sequence_length=30, augmentation_factor=3)

# Generate sequences from our static landmarks
print("Generating temporal sequences from static landmarks...")
sequences, sequence_labels = sequence_generator.generate_sequences(features, labels)

print(f"\nSequence generation completed!")
print(f"Sequence Statistics:")
print(f"  - Original static landmarks: {len(features):,}")
print(f"  - Generated sequences: {len(sequences):,}")
print(f"  - Sequence shape: {sequences.shape}")
print(f"  - Augmentation factor: {sequence_generator.augmentation_factor}")
print(f"  - Sequence length: {sequence_generator.sequence_length} frames")

# Visualize sequence characteristics
print("\nVisualizing sequence characteristics...")

# Select a sample sequence to visualize
sample_idx = 0
sample_sequence = sequences[sample_idx]
sample_label = sequence_labels[sample_idx]

# Plot landmark variation over time
plt.figure(figsize=(15, 8))

# Plot first 6 landmark coordinates over time
landmark_indices = [0, 2, 4, 6, 8, 10]  # x-coordinates of first 6 landmarks
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown']

for i, (landmark_idx, color) in enumerate(zip(landmark_indices, colors)):
    landmark_values = sample_sequence[:, landmark_idx]
    plt.subplot(2, 3, i + 1)
    plt.plot(landmark_values, color=color, linewidth=2)
    plt.title(f'Landmark {landmark_idx//2} (x-coord) - Class {sample_label}', fontweight='bold')
    plt.xlabel('Frame')
    plt.ylabel('Normalized Coordinate')
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Show class distribution in sequences
unique_labels, counts = np.unique(sequence_labels, return_counts=True)
print(f"\nSequence class distribution:")
for label, count in zip(unique_labels, counts):
    print(f"  {label}: {count:,} sequences")

# Save sequences for training
sequences_output_path = os.path.join(output_path, 'sequences.npz')
np.savez_compressed(sequences_output_path, sequences=sequences, labels=sequence_labels)
print(f"\nSequences saved to {sequences_output_path}")
````

````python
# Cell 8: Data Preprocessing for LSTM Training
"""
Preparing our sequence data for LSTM training through comprehensive preprocessing.

Preprocessing is crucial for LSTM performance and training stability.
"""

class ASLDataPreprocessor:
    """
    Comprehensive data preprocessing pipeline for ASL sequence data.
    
    Handles:
    - Label encoding and categorical conversion
    - Train/validation/test splitting with stratification
    - Feature normalization for stable training
    - Data format conversion for TensorFlow
    """
    
    def __init__(self, test_size=0.2, validation_size=0.2, random_state=42):
        """
        Initialize the preprocessor with split configuration.
        
        Args:
            test_size: Proportion of data for testing
            validation_size: Proportion of training data for validation
            random_state: Random seed for reproducible splits
        """
        self.test_size = test_size
        self.validation_size = validation_size
        self.random_state = random_state
        
        # Initialize encoders and scalers
        self.label_encoder = LabelEncoder()
        self.feature_stats = {}
        
    def prepare_dataset(self, sequences, labels):
        """
        Complete preprocessing pipeline for sequence data.
        
        Args:
            sequences: Input sequence data [samples, time_steps, features]
            labels: Corresponding string labels
            
        Returns:
            Tuple of preprocessed datasets and metadata
        """
        print("PREPROCESSING ASL SEQUENCE DATA")
        print("="*40)
        
        # Step 1: Encode labels
        print("Step 1: Encoding labels...")
        y_encoded = self.label_encoder.fit_transform(labels)
        y_categorical = tf.keras.utils.to_categorical(y_encoded)
        
        num_classes = len(self.label_encoder.classes_)
        print(f"  Encoded {num_classes} classes")
        print(f"  Classes: {list(self.label_encoder.classes_)}")
        
        # Step 2: Split data into train/validation/test
        print("\nStep 2: Splitting data...")
        
        # First split: separate test set
        X_temp, X_test, y_temp, y_test = train_test_split(
            sequences, y_categorical,
            test_size=self.test_size,
            stratify=y_encoded,
            random_state=self.random_state
        )
        
        # Second split: separate training and validation
        X_train, X_val, y_train, y_val = train_test_split(
            X_temp, y_temp,
            test_size=self.validation_size,
            stratify=np.argmax(y_temp, axis=1),
            random_state=self.random_state
        )
        
        print(f"   Training set: {X_train.shape[0]:,} sequences")
        print(f"   Validation set: {X_val.shape[0]:,} sequences")
        print(f"   Test set: {X_test.shape[0]:,} sequences")
        
        # Step 3: Normalize features
        print("\nStep 3: Normalizing features...")
        X_train_norm = self._normalize_sequences(X_train, fit=True)
        X_val_norm = self._normalize_sequences(X_val, fit=False)
        X_test_norm = self._normalize_sequences(X_test, fit=False)
        
        print(f"   Applied normalization with μ={np.mean(self.feature_stats['mean']):.4f}, σ={np.mean(self.feature_stats['std']):.4f}")
        
        # Step 4: Data quality checks
        print("\nStep 4: Data quality validation...")
        self._validate_data_quality(X_train_norm, y_train, "Training")
        self._validate_data_quality(X_val_norm, y_val, "Validation")
        self._validate_data_quality(X_test_norm, y_test, "Test")
        
        return {
            'X_train': X_train_norm,
            'X_val': X_val_norm,
            'X_test': X_test_norm,
            'y_train': y_train,
            'y_val': y_val,
            'y_test': y_test,
            'num_classes': num_classes,
            'input_shape': X_train_norm.shape[1:],
            'class_names': self.label_encoder.classes_
        }
    
    def _normalize_sequences(self, sequences, fit=False):
        """
        Normalize sequence features for stable training.
        
        Args:
            sequences: Input sequences
            fit: Whether to compute normalization statistics
            
        Returns:
            Normalized sequences
        """
        original_shape = sequences.shape
        
        # Reshape to (samples * time_steps, features) for normalization
        reshaped = sequences.reshape(-1, sequences.shape[-1])
        
        if fit:
            # Compute normalization statistics
            self.feature_stats['mean'] = np.mean(reshaped, axis=0)
            self.feature_stats['std'] = np.std(reshaped, axis=0) + 1e-8  # Avoid division by zero
            
            print(f"    Computed normalization statistics:")
            print(f"    Mean range: [{np.min(self.feature_stats['mean']):.4f}, {np.max(self.feature_stats['mean']):.4f}]")
            print(f"    Std range: [{np.min(self.feature_stats['std']):.4f}, {np.max(self.feature_stats['std']):.4f}]")
        
        # Apply normalization
        normalized = (reshaped - self.feature_stats['mean']) / self.feature_stats['std']
        
        # Reshape back to original dimensions
        return normalized.reshape(original_shape)
    
    def _validate_data_quality(self, X, y, split_name):
        """
        Validate data quality and check for potential issues.
        
        Args:
            X: Input sequences
            y: Target labels
            split_name: Name of the data split for logging
        """
        # Check for NaN or infinite values
        nan_count = np.sum(np.isnan(X))
        inf_count = np.sum(np.isinf(X))
        
        if nan_count > 0:
            print(f"     {split_name}: Found {nan_count} NaN values")
        
        if inf_count > 0:
            print(f"     {split_name}: Found {inf_count} infinite values")
        
        if nan_count == 0 and inf_count == 0:
            print(f"    {split_name}: Data quality validated")
        
        # Check class distribution
        class_counts = np.sum(y, axis=0)
        min_class_count = np.min(class_counts)
        max_class_count = np.max(class_counts)
        
        if max_class_count / min_class_count > 2:
            print(f"    {split_name}: Class imbalance detected (ratio: {max_class_count/min_class_count:.2f})")
        else:
            print(f"    {split_name}: Class distribution balanced")
    
    def decode_prediction(self, prediction):
        """
        Decode model prediction back to class label.
        
        Args:
            prediction: Model prediction probabilities
            
        Returns:
            Tuple of (predicted_label, confidence)
        """
        class_idx = np.argmax(prediction)
        confidence = prediction[class_idx]
        label = self.label_encoder.inverse_transform([class_idx])[0]
        
        return label, confidence
    
    def get_class_weights(self, y):
        """
        Calculate class weights for handling class imbalance.
        
        Args:
            y: One-hot encoded labels
            
        Returns:
            Dictionary of class weights
        """
        from sklearn.utils.class_weight import compute_class_weight
        
        # Convert one-hot to class indices
        y_indices = np.argmax(y, axis=1)
        
        # Compute class weights
        class_weights = compute_class_weight(
            'balanced',
            classes=np.unique(y_indices),
            y=y_indices
        )
        
        return dict(enumerate(class_weights))

# Initialize preprocessor and process our data
print("Initializing data preprocessor...")
preprocessor = ASLDataPreprocessor(test_size=0.2, validation_size=0.2, random_state=42)

# Process the sequence data
processed_data = preprocessor.prepare_dataset(sequences, sequence_labels)

print(f"\nDATA PREPROCESSING COMPLETED!")
print(f"Final Dataset Statistics:")
print(f"  - Input shape: {processed_data['input_shape']}")
print(f"  - Number of classes: {processed_data['num_classes']}")
print(f"  - Training samples: {processed_data['X_train'].shape[0]:,}")
print(f"  - Validation samples: {processed_data['X_val'].shape[0]:,}")
print(f"  - Test samples: {processed_data['X_test'].shape[0]:,}")
print(f"  - Total sequences: {sum([processed_data['X_train'].shape[0], processed_data['X_val'].shape[0], processed_data['X_test'].shape[0]]):,}")

# Calculate and display class weights for handling any imbalance
class_weights = preprocessor.get_class_weights(processed_data['y_train'])
print(f"\n Class weights for balanced training:")
for i, (class_idx, weight) in enumerate(class_weights.items()):
    class_name = processed_data['class_names'][class_idx]
    print(f"  {class_name}: {weight:.3f}")
````

````python
# Cell 9: LSTM Model Architecture Design
"""
Designing and implementing our LSTM neural network architecture for ASL recognition.
"""

def create_advanced_lstm_model(input_shape, num_classes, 
                              lstm_units=[128, 64], 
                              dense_units=[64, 32],
                              dropout_rate=0.3,
                              learning_rate=0.001,
                              use_attention=False):
    """
    Create an advanced LSTM model for ASL recognition.
    
    This architecture incorporates:
    - Multi-layer LSTM with different unit sizes
    - Batch normalization for training stability
    - Dropout for regularization
    - Dense layers for classification
    - Optional attention mechanism
    
    Args:
        input_shape: Shape of input sequences (time_steps, features)
        num_classes: Number of output classes
        lstm_units: List of LSTM layer sizes
        dense_units: List of dense layer sizes
        dropout_rate: Dropout rate for regularization
        learning_rate: Learning rate for optimizer
        use_attention: Whether to include attention mechanism
        
    Returns:
        Compiled Keras model
    """
    print("BUILDING ADVANCED LSTM ARCHITECTURE")
    print("="*45)
    
    model = Sequential()
    
    # First LSTM layer - captures initial temporal patterns
    model.add(LSTM(
        lstm_units[0],          # 128 memory cells
        return_sequences=True,  # Return sequences for next LSTM layer
        input_shape=input_shape, # 30 timesteps. 42 features per frame
        kernel_regularizer=tf.keras.regularizers.l2(0.001),
        recurrent_regularizer=tf.keras.regularizers.l2(0.001),
        name='lstm_1'
    ))
    model.add(BatchNormalization(name='bn_1'))
    model.add(Dropout(dropout_rate, name='dropout_1'))
    
    # Second LSTM layer - captures higher-level temporal patterns
    model.add(LSTM(
        lstm_units[1],           # 64 memory cells
        return_sequences=False,  # Don't return sequences for classification
        kernel_regularizer=tf.keras.regularizers.l2(0.001),
        recurrent_regularizer=tf.keras.regularizers.l2(0.001),
        name='lstm_2'
    ))
    model.add(BatchNormalization(name='bn_2'))
    model.add(Dropout(dropout_rate, name='dropout_2'))
    
    # Dense layers for classification
    for i, units in enumerate(dense_units):
        model.add(Dense(
            units,
            activation='relu',
            kernel_regularizer=tf.keras.regularizers.l2(0.001),
            name=f'dense_{i+1}'
        ))
        model.add(BatchNormalization(name=f'bn_dense_{i+1}'))
        
        # Reduce dropout in later layers
        dropout_factor = 0.5 if i == len(dense_units) - 1 else 1.0
        model.add(Dropout(dropout_rate * dropout_factor, name=f'dropout_dense_{i+1}'))
    
    # Output layer with softmax activation
    model.add(Dense(
        num_classes,
        activation='softmax',
        name='output'
    ))
    
    # Compile model with advanced optimizer and metrics
    optimizer = Adam(
        learning_rate=learning_rate,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-7
    )
    
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.TopKCategoricalAccuracy(k=3, name='top_3_accuracy'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )
    
    return model

# Create our ASL recognition model
print("Creating ASL recognition model...")

# Define model architecture parameters
model_config = {
    'input_shape': processed_data['input_shape'],
    'num_classes': processed_data['num_classes'],
    'lstm_units': [128, 64],
    'dense_units': [64, 32],
    'dropout_rate': 0.3,
    'learning_rate': 0.001,
    'use_attention': False
}

# Build the model
asl_model = create_advanced_lstm_model(**model_config)

# Display model architecture
print("\nMODEL ARCHITECTURE SUMMARY")
print("="*40)
asl_model.summary()

# Calculate and display model complexity
total_params = asl_model.count_params()
trainable_params = sum([tf.keras.backend.count_params(w) for w in asl_model.trainable_weights])
non_trainable_params = total_params - trainable_params

print(f"\nMODEL COMPLEXITY:")
print(f"  - Total parameters: {total_params:,}")
print(f"  - Trainable parameters: {trainable_params:,}")
print(f"  - Non-trainable parameters: {non_trainable_params:,}")
print(f"  - Model size (approx): {total_params * 4 / 1024 / 1024:.2f} MB")

# Visualize model architecture
print("\nVisualizing model architecture...")
try:
    tf.keras.utils.plot_model(
        asl_model,
        to_file='asl_model_architecture.png',
        show_shapes=True,
        show_layer_names=True,
        rankdir='TB',
        dpi=150
    )
    print("Model architecture diagram saved to asl_model_architecture.png")
except Exception as e:
    print(f"Note: Could not create model diagram - {e}")

# Analyze model capacity and complexity
print(f"\nMODEL ANALYSIS:")
print(f"  - Input sequence length: {model_config['input_shape'][0]} frames")
print(f"  - Input features per frame: {model_config['input_shape'][1]}")
print(f"  - LSTM memory cells: {sum(model_config['lstm_units'])}")
print(f"  - Classification classes: {model_config['num_classes']}")
print(f"  - Regularization: L2 + Dropout ({model_config['dropout_rate']})")
print(f"  - Optimizer: Adam (lr={model_config['learning_rate']})")
````

````python
# Cell 10: Model Training with Advanced Callbacks
"""
Training our LSTM model with comprehensive monitoring and optimization strategies.
"""

def create_training_callbacks(model_name='best_asl_model', patience=15):
    """
    Create comprehensive training callbacks for optimal training.
    
    Args:
        model_name: Name for saved model checkpoints
        patience: Patience for early stopping
        
    Returns:
        List of configured callbacks
    """
    callbacks = [
        # Early stopping to prevent overfitting
        EarlyStopping(
            monitor='val_accuracy',
            patience=patience,
            restore_best_weights=True,
            verbose=1,
            mode='max'
        ),
        
        # Reduce learning rate when validation accuracy plateaus
        ReduceLROnPlateau(
            monitor='val_accuracy',
            factor=0.5,
            patience=patience//2,
            min_lr=1e-7,
            verbose=1,
            mode='max'
        ),
        
        # Save best model weights
        ModelCheckpoint(
            filepath=f'{model_name}.h5',
            monitor='val_accuracy',
            save_best_only=True,
            save_weights_only=False,
            verbose=1,
            mode='max'
        ),
        
        # Custom callback for detailed logging
        tf.keras.callbacks.CSVLogger(
            filename=f'{model_name}_training_log.csv',
            separator=',',
            append=False
        )
    ]
    
    return callbacks

def train_asl_model(model, train_data, validation_data, 
                   epochs=100, batch_size=32, 
                   class_weights=None, verbose=1):
    """
    Train the ASL recognition model with comprehensive monitoring.
    
    Args:
        model: Compiled Keras model
        train_data: Tuple of (X_train, y_train)
        validation_data: Tuple of (X_val, y_val)
        epochs: Maximum number of training epochs
        batch_size: Training batch size
        class_weights: Dictionary of class weights for imbalanced data
        verbose: Verbosity level
        
    Returns:
        Training history object
    """
    print("TRAINING ASL RECOGNITION MODEL")
    print("="*40)
    
    X_train, y_train = train_data
    X_val, y_val = validation_data
    
    print(f"Training Configuration:")
    print(f"  - Training samples: {X_train.shape[0]:,}")
    print(f"  - Validation samples: {X_val.shape[0]:,}")
    print(f"  - Batch size: {batch_size}")
    print(f"  - Max epochs: {epochs}")
    print(f"  - Class weights: {'Yes' if class_weights else 'No'}")
    
    # Create callbacks
    callbacks = create_training_callbacks(model_name='best_asl_model', patience=15)
    
    # Start training
    print(f"\nStarting training...")
    start_time = time.time()
    
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=callbacks,
        class_weight=class_weights,
        verbose=verbose,
        shuffle=True
    )
    
    training_time = time.time() - start_time
    
    print(f"\nTraining completed!")
    print(f"  - Total training time: {training_time:.2f} seconds")
    print(f"  - Average time per epoch: {training_time/len(history.history['loss']):.2f} seconds")
    print(f"  - Final training accuracy: {history.history['accuracy'][-1]:.4f}")
    print(f"  - Final validation accuracy: {history.history['val_accuracy'][-1]:.4f}")
    print(f"  - Best validation accuracy: {max(history.history['val_accuracy']):.4f}")
    
    return history

# Prepare training data
print("Preparing training data...")
train_data = (processed_data['X_train'], processed_data['y_train'])
validation_data = (processed_data['X_val'], processed_data['y_val'])

# Calculate class weights to handle any imbalance
class_weights = preprocessor.get_class_weights(processed_data['y_train'])

# Train the model
training_history = train_asl_model(
    model=asl_model,
    train_data=train_data,
    validation_data=validation_data,
    epochs=50,  # Reduced for demonstration, increase for better results
    batch_size=32,
    class_weights=class_weights,
    verbose=1
)
````

````python
# Cell 11: Training Visualization and Analysis
"""
Comprehensive visualization and analysis of the training process for our ASL recognition model.

- Training and validation curves
- Loss progression analysis
- Learning rate optimization effects
- Overfitting detection and analysis
- Training stability metrics
"""

def create_comprehensive_training_analysis(training_history, model_config=None):
    """
    Create comprehensive training analysis visualizations.
    
    Args:
        training_history: Keras training history object
        model_config: Dictionary containing model configuration details
        
    Returns:
        Dictionary containing analysis results
    """
    print("COMPREHENSIVE TRAINING ANALYSIS")
    print("=" * 50)
    
    # Extract training history data
    history = training_history.history
    epochs = range(1, len(history['accuracy']) + 1)
    
    # Create comprehensive training visualization
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('ASL LSTM Model: Complete Training Analysis', fontsize=16, fontweight='bold')
    
    # 1. Accuracy Progression
    ax1 = axes[0, 0]
    ax1.plot(epochs, history['accuracy'], 'b-', label='Training Accuracy', linewidth=2, marker='o', markersize=4)
    ax1.plot(epochs, history['val_accuracy'], 'r--', label='Validation Accuracy', linewidth=2, marker='s', markersize=4)
    
    # Highlight best validation epoch
    best_val_epoch = np.argmax(history['val_accuracy']) + 1
    best_val_acc = max(history['val_accuracy'])
    ax1.axvline(x=best_val_epoch, color='green', linestyle=':', alpha=0.7, linewidth=2)
    ax1.annotate(f'Best Val: {best_val_acc:.4f}\nEpoch {best_val_epoch}', 
                xy=(best_val_epoch, best_val_acc), xytext=(best_val_epoch + 3, best_val_acc - 0.05),
                arrowprops=dict(arrowstyle='->', color='green', lw=1.5),
                bbox=dict(boxstyle="round,pad=0.3", facecolor='lightgreen', alpha=0.7),
                fontsize=9, fontweight='bold')
    
    ax1.set_title('Model Accuracy Over Time', fontweight='bold', fontsize=12)
    ax1.set_xlabel('Epoch', fontweight='bold')
    ax1.set_ylabel('Accuracy', fontweight='bold')
    ax1.legend(loc='lower right')
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(0, 1.05)
    
    # 2. Loss Progression
    ax2 = axes[0, 1]
    ax2.plot(epochs, history['loss'], 'b-', label='Training Loss', linewidth=2, marker='o', markersize=4)
    ax2.plot(epochs, history['val_loss'], 'r--', label='Validation Loss', linewidth=2, marker='s', markersize=4)
    
    # Highlight minimum validation loss
    best_loss_epoch = np.argmin(history['val_loss']) + 1
    best_val_loss = min(history['val_loss'])
    ax2.axvline(x=best_loss_epoch, color='orange', linestyle=':', alpha=0.7, linewidth=2)
    ax2.annotate(f'Min Val Loss: {best_val_loss:.4f}\nEpoch {best_loss_epoch}', 
                xy=(best_loss_epoch, best_val_loss), xytext=(best_loss_epoch + 3, best_val_loss + 0.1),
                arrowprops=dict(arrowstyle='->', color='orange', lw=1.5),
                bbox=dict(boxstyle="round,pad=0.3", facecolor='lightyellow', alpha=0.7),
                fontsize=9, fontweight='bold')
    
    ax2.set_title('Model Loss Over Time', fontweight='bold', fontsize=12)
    ax2.set_xlabel('Epoch', fontweight='bold')
    ax2.set_ylabel('Loss', fontweight='bold')
    ax2.legend(loc='upper right')
    ax2.grid(True, alpha=0.3)
    ax2.set_yscale('log')  # Log scale for better loss visualization
    
    # 3. Learning Rate Effects (if available in history)
    ax3 = axes[0, 2]
    if 'lr' in history:
        # Plot learning rate changes
        ax3_twin = ax3.twinx()
        
        line1 = ax3.plot(epochs, history['val_accuracy'], 'g-', label='Validation Accuracy', linewidth=2)
        line2 = ax3_twin.plot(epochs, history['lr'], 'purple', label='Learning Rate', linewidth=2, alpha=0.7)
        
        ax3.set_xlabel('Epoch', fontweight='bold')
        ax3.set_ylabel('Validation Accuracy', color='g', fontweight='bold')
        ax3_twin.set_ylabel('Learning Rate', color='purple', fontweight='bold')
        ax3.set_title('Learning Rate Schedule Impact', fontweight='bold', fontsize=12)
        ax3.grid(True, alpha=0.3)
        
        # Combined legend
        lines = line1 + line2
        labels = [l.get_label() for l in lines]
        ax3.legend(lines, labels, loc='center right')
    else:
        # Alternative: show training stability metrics
        training_acc_smooth = np.convolve(history['accuracy'], np.ones(5)/5, mode='valid')
        val_acc_smooth = np.convolve(history['val_accuracy'], np.ones(5)/5, mode='valid')
        smooth_epochs = range(3, len(history['accuracy']) - 1)
        
        ax3.plot(smooth_epochs, training_acc_smooth, 'b-', label='Training (smoothed)', linewidth=3, alpha=0.8)
        ax3.plot(smooth_epochs, val_acc_smooth, 'r-', label='Validation (smoothed)', linewidth=3, alpha=0.8)
        ax3.set_title('Training Stability (5-Epoch Moving Average)', fontweight='bold', fontsize=12)
        ax3.set_xlabel('Epoch', fontweight='bold')
        ax3.set_ylabel('Accuracy', fontweight='bold')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
    
    # 4. Generalization Gap Analysis
    ax4 = axes[1, 0]
    generalization_gap = np.array(history['accuracy']) - np.array(history['val_accuracy'])
    
    ax4.plot(epochs, generalization_gap, 'purple', linewidth=2, marker='d', markersize=4)
    ax4.axhline(y=0, color='black', linestyle='-', alpha=0.5, linewidth=1)
    ax4.axhline(y=0.05, color='red', linestyle='--', alpha=0.7, linewidth=1, label='Overfitting Threshold')
    ax4.axhline(y=-0.05, color='blue', linestyle='--', alpha=0.7, linewidth=1, label='Underfitting Threshold')
    
    # Color-code regions
    ax4.fill_between(epochs, 0.05, 1, alpha=0.2, color='red', label='Overfitting Zone')
    ax4.fill_between(epochs, -1, -0.05, alpha=0.2, color='blue', label='Underfitting Zone')
    ax4.fill_between(epochs, -0.05, 0.05, alpha=0.2, color='green', label='Good Generalization')
    
    ax4.set_title('Generalization Gap Analysis', fontweight='bold', fontsize=12)
    ax4.set_xlabel('Epoch', fontweight='bold')
    ax4.set_ylabel('Training - Validation Accuracy', fontweight='bold')
    ax4.legend(loc='upper right', fontsize=9)
    ax4.grid(True, alpha=0.3)
    
    # 5. Additional Training Metrics (if available)
    ax5 = axes[1, 1]
    if 'precision' in history and 'recall' in history:
        ax5.plot(epochs, history['precision'], 'orange', label='Training Precision', linewidth=2, marker='^', markersize=4)
        ax5.plot(epochs, history['recall'], 'cyan', label='Training Recall', linewidth=2, marker='v', markersize=4)
        
        if 'val_precision' in history and 'val_recall' in history:
            ax5.plot(epochs, history['val_precision'], 'red', label='Val Precision', linewidth=2, linestyle='--', marker='^', markersize=4)
            ax5.plot(epochs, history['val_recall'], 'blue', label='Val Recall', linewidth=2, linestyle='--', marker='v', markersize=4)
        
        ax5.set_title('Precision & Recall Progression', fontweight='bold', fontsize=12)
        ax5.set_xlabel('Epoch', fontweight='bold')
        ax5.set_ylabel('Metric Value', fontweight='bold')
        ax5.legend(fontsize=9)
        ax5.grid(True, alpha=0.3)
        ax5.set_ylim(0, 1.05)
    else:
        # Show loss components breakdown
        ax5.plot(epochs, history['loss'], 'red', label='Total Loss', linewidth=3)
        ax5.plot(epochs, history['val_loss'], 'blue', label='Validation Loss', linewidth=3, linestyle='--')
        ax5.set_title('Loss Components Analysis', fontweight='bold', fontsize=12)
        ax5.set_xlabel('Epoch', fontweight='bold')
        ax5.set_ylabel('Loss Value', fontweight='bold')
        ax5.legend()
        ax5.grid(True, alpha=0.3)
        ax5.set_yscale('log')
    
    # 6. Training Summary Statistics
    ax6 = axes[1, 2]
    ax6.axis('off')
    
    # Calculate comprehensive training statistics
    final_train_acc = history['accuracy'][-1]
    final_val_acc = history['val_accuracy'][-1]
    best_val_acc = max(history['val_accuracy'])
    final_generalization_gap = final_train_acc - final_val_acc
    
    # Training stability (variance in last 10 epochs)
    last_10_epochs = min(10, len(history['val_accuracy']))
    val_acc_stability = np.std(history['val_accuracy'][-last_10_epochs:])
    
    # Convergence analysis
    early_acc = np.mean(history['val_accuracy'][:5]) if len(history['val_accuracy']) >= 5 else history['val_accuracy'][0]
    improvement_rate = (best_val_acc - early_acc) / len(epochs)
    
    # Training efficiency
    epochs_to_90_percent = None
    for i, acc in enumerate(history['val_accuracy']):
        if acc >= 0.9 * best_val_acc:
            epochs_to_90_percent = i + 1
            break
    
    summary_text = f"""TRAINING SUMMARY STATISTICS
    
Final Performance:
• Training Accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)
• Validation Accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)
• Best Validation Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%)

Generalization Analysis:
• Final Generalization Gap: {final_generalization_gap:.4f}
• Status: {'Good' if abs(final_generalization_gap) < 0.05 else 'Overfitting' if final_generalization_gap > 0.05 else 'Underfitting'}
• Validation Stability (σ): {val_acc_stability:.4f}

Training Efficiency:
• Total Epochs: {len(epochs)}
• Best Epoch: {best_val_epoch}
• Improvement Rate: {improvement_rate:.4f}/epoch
• Epochs to 90% Best: {epochs_to_90_percent if epochs_to_90_percent else 'N/A'}

Model Complexity:
• Architecture: {'LSTM-based' if model_config else 'Neural Network'}
• Parameters: {model_config.get('total_parameters', 'N/A') if model_config else 'N/A'}
• Optimization: {'Adam with LR scheduling' if 'lr' in history else 'Adam optimizer'}
"""
    
    ax6.text(0.05, 0.95, summary_text, transform=ax6.transAxes, fontsize=10,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle="round,pad=0.5", facecolor='lightgray', alpha=0.8))
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    
    # Save the comprehensive analysis
    plt.savefig('training_analysis_comprehensive.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Compile analysis results
    analysis_results = {
        'final_train_accuracy': final_train_acc,
        'final_val_accuracy': final_val_acc,
        'best_val_accuracy': best_val_acc,
        'best_epoch': best_val_epoch,
        'generalization_gap': final_generalization_gap,
        'training_stability': val_acc_stability,
        'improvement_rate': improvement_rate,
        'epochs_to_90_percent': epochs_to_90_percent,
        'total_epochs': len(epochs),
        'overfitting_detected': final_generalization_gap > 0.05,
        'training_converged': val_acc_stability < 0.01
    }
    
    return analysis_results

def analyze_training_patterns(training_history):
    """
    Analyze specific training patterns and provide recommendations.
    
    Args:
        training_history: Keras training history object
        
    Returns:
        Dictionary containing pattern analysis and recommendations
    """
    history = training_history.history
    
    print("\nTRAINING PATTERN ANALYSIS")
    print("=" * 40)
    
    # Pattern detection
    patterns = {
        'overfitting': False,
        'underfitting': False,
        'unstable_training': False,
        'early_plateau': False,
        'late_improvement': False
    }
    
    recommendations = []
    
    # Check for overfitting
    final_gap = history['accuracy'][-1] - history['val_accuracy'][-1]
    if final_gap > 0.05:
        patterns['overfitting'] = True
        recommendations.append("Consider increasing dropout rate or reducing model complexity")
        recommendations.append("Add more training data or data augmentation")
        recommendations.append("Implement early stopping with patience")
    
    # Check for underfitting
    if max(history['val_accuracy']) < 0.8:
        patterns['underfitting'] = True
        recommendations.append("Increase model complexity (more LSTM units or layers)")
        recommendations.append("Reduce dropout rate")
        recommendations.append("Train for more epochs")
        recommendations.append("Check data quality and preprocessing")
    
    # Check for training stability
    val_acc_variations = np.diff(history['val_accuracy'])
    if np.std(val_acc_variations) > 0.05:
        patterns['unstable_training'] = True
        recommendations.append("Reduce learning rate for training")
        recommendations.append("Implement learning rate scheduling")
        recommendations.append("Check batch size - consider smaller batches")
    
    # Check for early plateau
    if len(history['val_accuracy']) > 10:
        early_plateau_check = np.std(history['val_accuracy'][5:15]) < 0.01
        if early_plateau_check:
            patterns['early_plateau'] = True
            recommendations.append("Consider learning rate warm-up")
            recommendations.append("Try different initialization strategies")
            recommendations.append("Implement cyclical learning rates")
    
    # Check for late improvement
    if len(history['val_accuracy']) > 20:
        early_acc = np.mean(history['val_accuracy'][:10])
        late_acc = np.mean(history['val_accuracy'][-10:])
        if late_acc - early_acc > 0.1:
            patterns['late_improvement'] = True
            recommendations.append("Training shows good convergence - current settings work well")
            recommendations.append("Consider training for more epochs if computational budget allows")
    
    # Display findings
    print("Pattern Detection Results:")
    for pattern, detected in patterns.items():
        status = "✓ DETECTED" if detected else "✗ Not detected"
        print(f"  {pattern.replace('_', ' ').title()}: {status}")
    
    if recommendations:
        print(f"\nRecommendations for improvement:")
        for i, rec in enumerate(recommendations, 1):
            print(f"  {i}. {rec}")
    else:
        print(f"\nNo significant issues detected. Training appears well-configured.")
    
    return {
        'patterns': patterns,
        'recommendations': recommendations,
        'analysis_summary': f"Detected {sum(patterns.values())} training patterns requiring attention"
    }

# Perform comprehensive training analysis
print("Analyzing training results and generating comprehensive visualizations...")

# Check if training history is available
if 'training_history' in globals():
    print("Training history found - proceeding with analysis...")
    
    # Create comprehensive training analysis
    training_analysis = create_comprehensive_training_analysis(
        training_history, 
        model_config={'total_parameters': asl_model.count_params() if 'asl_model' in globals() else None}
    )
    
    # Analyze training patterns and provide recommendations
    pattern_analysis = analyze_training_patterns(training_history)
    
    print(f"\nTRAINING ANALYSIS COMPLETED")
    print(f"=" * 40)
    print(f"Key Findings:")
    print(f"  • Final validation accuracy: {training_analysis['final_val_accuracy']:.4f}")
    print(f"  • Best validation accuracy: {training_analysis['best_val_accuracy']:.4f}")
    print(f"  • Training converged: {'Yes' if training_analysis['training_converged'] else 'No'}")
    print(f"  • Overfitting detected: {'Yes' if training_analysis['overfitting_detected'] else 'No'}")
    print(f"  • Training stability: {training_analysis['training_stability']:.4f}")
    
    # Save analysis results
    analysis_results = {
        'training_metrics': training_analysis,
        'pattern_analysis': pattern_analysis,
        'timestamp': time.strftime('%Y-%m-%d %H:%M:%S')
    }
    
    # Save to file for later reference
    with open('training_analysis_results.pkl', 'wb') as f:
        pickle.dump(analysis_results, f)
    
    print(f"\nAnalysis results saved to 'training_analysis_results.pkl'")
    print(f"Visualization saved to 'training_analysis_comprehensive.png'")
    
else:
    print("Training history not found. Please ensure model training (Cell 10) has been completed.")
    print("If you're running cells individually, make sure to run Cell 10 first.")
````

````python
# Cell 12: Model Evaluation and Testing
"""
Comprehensive evaluation of our trained ASL recognition model.
"""

def evaluate_asl_model(model, test_data, class_names, detailed_analysis=True):
    """
    Comprehensive evaluation of the trained ASL model.
    
    Args:
        model: Trained Keras model
        test_data: Tuple of (X_test, y_test)
        class_names: List of class names
        detailed_analysis: Whether to perform detailed analysis
        
    Returns:
        Dictionary containing evaluation results
    """
    print("COMPREHENSIVE MODEL EVALUATION")
    print("="*40)
    
    X_test, y_test = test_data
    
    # Generate predictions
    print("Generating predictions on test set...")
    start_time = time.time()
    y_pred_proba = model.predict(X_test, verbose=0)
    prediction_time = time.time() - start_time
    
    y_pred = np.argmax(y_pred_proba, axis=1)
    y_true = np.argmax(y_test, axis=1)
    
    # Calculate basic metrics
    test_loss, test_accuracy, test_top3_acc, test_precision, test_recall = model.evaluate(
        X_test, y_test, verbose=0
    )
    
    # Calculate additional metrics
    from sklearn.metrics import f1_score, precision_score, recall_score
    
    # Weighted averages account for class imbalance
    precision_weighted = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    recall_weighted = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1_weighted = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    
    # Macro averages treat all classes equally
    precision_macro = precision_score(y_true, y_pred, average='macro', zero_division=0)
    recall_macro = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1_macro = f1_score(y_true, y_pred, average='macro', zero_division=0)
    
    # Performance metrics
    avg_prediction_time = prediction_time / len(X_test) * 1000  # ms per sample
    
    # Compile results
    results = {
        'test_loss': test_loss,
        'test_accuracy': test_accuracy,
        'test_top3_accuracy': test_top3_acc,
        'test_precision': test_precision,
        'test_recall': test_recall,
        'precision_weighted': precision_weighted,
        'recall_weighted': recall_weighted,
        'f1_weighted': f1_weighted,
        'precision_macro': precision_macro,
        'recall_macro': recall_macro,
        'f1_macro': f1_macro,
        'avg_prediction_time_ms': avg_prediction_time,
        'total_test_samples': len(X_test),
        'predictions': y_pred,
        'true_labels': y_true,
        'prediction_probabilities': y_pred_proba
    }
    
    # Display results
    print(f"\nEVALUATION RESULTS")
    print("="*30)
    print(f"Overall Performance:")
    print(f"  - Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"  - Test Loss: {test_loss:.4f}")
    print(f"  - Top-3 Accuracy: {test_top3_acc:.4f} ({test_top3_acc*100:.2f}%)")
    
    print(f"\nDetailed Metrics:")
    print(f"  - Precision (weighted): {precision_weighted:.4f}")
    print(f"  - Recall (weighted): {recall_weighted:.4f}")
    print(f"  - F1-Score (weighted): {f1_weighted:.4f}")
    print(f"  - Precision (macro): {precision_macro:.4f}")
    print(f"  - Recall (macro): {recall_macro:.4f}")
    print(f"  - F1-Score (macro): {f1_macro:.4f}")
    
    print(f"\nPerformance Metrics:")
    print(f"  - Average prediction time: {avg_prediction_time:.2f} ms")
    print(f"  - Total test samples: {len(X_test):,}")
    print(f"  - Throughput: {1000/avg_prediction_time:.1f} samples/second")
    
    if detailed_analysis:
        # Confusion Matrix Analysis
        print(f"\nDetailed Analysis:")
        
        # Get unique classes present in test data
        unique_true = np.unique(y_true)
        unique_pred = np.unique(y_pred)
        unique_classes = np.unique(np.concatenate([unique_true, unique_pred]))
        
        print(f"Classes in test data: {len(unique_true)}")
        print(f"Classes in predictions: {len(unique_pred)}")
        print(f"Total unique classes: {len(unique_classes)}")
        
        # Generate confusion matrix with explicit labels
        cm = confusion_matrix(y_true, y_pred, labels=unique_classes)
        
        # Filter class names to match the classes actually present
        present_class_names = [class_names[i] for i in unique_classes if i < len(class_names)]
        
        print(f"Confusion matrix shape: {cm.shape}")
        print(f"Present class names: {len(present_class_names)}")
        
        # Create detailed confusion matrix visualization
        plt.figure(figsize=(12, 10))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=present_class_names, yticklabels=present_class_names)
        plt.title('Confusion Matrix - ASL Recognition', fontsize=16, fontweight='bold')
        plt.xlabel('Predicted Label', fontsize=12)
        plt.ylabel('True Label', fontsize=12)
        plt.xticks(rotation=45)
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
        plt.show()
        
        # Per-class accuracy analysis (only for classes with samples)
        class_accuracies = []
        class_names_with_samples = []
        
        for i, class_idx in enumerate(unique_classes):
            if i < cm.shape[0] and cm[i].sum() > 0:  # Only if class has samples
                accuracy = cm[i, i] / cm[i].sum()
                class_accuracies.append(accuracy)
                if class_idx < len(class_names):
                    class_names_with_samples.append(class_names[class_idx])
                else:
                    class_names_with_samples.append(f"Class_{class_idx}")
        
        # Create per-class accuracy visualization
        if class_accuracies:  # Only if we have accuracies to plot
            plt.figure(figsize=(15, 6))
            bars = plt.bar(class_names_with_samples, class_accuracies, 
                          color=plt.cm.viridis(np.array(class_accuracies)))
            plt.title('Per-Class Accuracy Analysis', fontsize=16, fontweight='bold')
            plt.xlabel('ASL Sign', fontsize=12)
            plt.ylabel('Accuracy', fontsize=12)
            plt.xticks(rotation=45)
            plt.ylim(0, 1)
            
            # Add value labels on bars
            for bar, acc in zip(bars, class_accuracies):
                plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                        f'{acc:.3f}', ha='center', va='bottom', fontsize=10)
            
            plt.tight_layout()
            plt.savefig('per_class_accuracy.png', dpi=300, bbox_inches='tight')
            plt.show()
            
            # Performance analysis
            if len(class_accuracies) >= 5:
                best_classes = np.argsort(class_accuracies)[-5:]
                worst_classes = np.argsort(class_accuracies)[:5]
                
                print(f"\nBest Performing Signs:")
                for idx in reversed(best_classes):
                    print(f"  • {class_names_with_samples[idx]}: {class_accuracies[idx]:.4f} ({class_accuracies[idx]*100:.1f}%)")
                
                print(f"\nChallenging Signs:")
                for idx in worst_classes:
                    print(f"  • {class_names_with_samples[idx]}: {class_accuracies[idx]:.4f} ({class_accuracies[idx]*100:.1f}%)")
            else:
                print(f"\nPer-Class Performance:")
                for name, acc in zip(class_names_with_samples, class_accuracies):
                    print(f"  • {name}: {acc:.4f} ({acc*100:.1f}%)")
        
        # Detailed classification report with proper labels
        print(f"\nCLASSIFICATION REPORT:")
        print("="*50)
        
        # Use only the classes that are actually present in the test set
        try:
            # Get labels that appear in both y_true and y_pred
            labels_in_data = sorted(list(set(y_true) | set(y_pred)))
            target_names_filtered = [class_names[i] if i < len(class_names) else f"Class_{i}" 
                                   for i in labels_in_data]
            
            report = classification_report(y_true, y_pred, 
                                         labels=labels_in_data,
                                         target_names=target_names_filtered,
                                         zero_division=0)
            print(report)
        except Exception as e:
            print(f"Could not generate detailed classification report: {e}")
            # Fallback: basic report without target names
            report = classification_report(y_true, y_pred, zero_division=0)
            print(report)
    
    return results

# Evaluate the trained model
print("Evaluating trained ASL recognition model...")
test_data = (processed_data['X_test'], processed_data['y_test'])
class_names = processed_data['class_names']

evaluation_results = evaluate_asl_model(
    model=asl_model,
    test_data=test_data,
    class_names=class_names,
    detailed_analysis=True
)
````

````python
### Cell 13: Real-Time Recognition System Implementation
"""
Implementing a complete real-time ASL recognition system that integrates
webcam input, MediaPipe hand tracking, and our trained LSTM model.
"""

class RealTimeASLRecognizer:
    """
    Real-time ASL recognition system with webcam integration.
    
    This class provides:
    - Real-time webcam processing
    - Hand landmark extraction
    - Sequence buffering for temporal analysis
    - LSTM model prediction
    - Confidence scoring and stability filtering
    - Visual feedback and performance monitoring
    """
    
    def __init__(self, model, preprocessor, sequence_length=30, confidence_threshold=0.7):
        """
        Initialize the real-time recognition system.
        
        Args:
            model: Trained LSTM model
            preprocessor: Data preprocessor with normalization parameters
            sequence_length: Number of frames to buffer for prediction
            confidence_threshold: Minimum confidence for accepting predictions
        """
        self.model = model
        self.preprocessor = preprocessor
        self.sequence_length = sequence_length
        self.confidence_threshold = confidence_threshold
        
        # Initialize MediaPipe hand extractor
        self.hand_extractor = HandLandmarkExtractor(
            static_image_mode=False,  # Video mode
            max_num_hands=1,
            min_detection_confidence=0.7,
            min_tracking_confidence=0.5
        )
        
        # Sequence buffer for temporal analysis
        self.sequence_buffer = deque(maxlen=sequence_length)
        
        # Performance tracking
        self.frame_times = deque(maxlen=30)
        self.prediction_times = deque(maxlen=30)
        
        # Prediction history for stability
        self.prediction_history = deque(maxlen=10)
        
        # Statistics
        self.total_frames = 0
        self.successful_predictions = 0
        
    def predict_from_sequence(self, landmarks_sequence):
        """
        Generate prediction from landmark sequence.
        
        Args:
            landmarks_sequence: Sequence of hand landmarks
            
        Returns:
            Tuple of (predicted_label, confidence, all_probabilities)
        """
        start_time = time.time()
        
        # Prepare sequence for model input
        sequence_array = np.array([landmarks_sequence])
        
        # Normalize using preprocessor statistics
        normalized_sequence = self.preprocessor._normalize_sequences(sequence_array, fit=False)
        
        # Generate prediction
        prediction_proba = self.model.predict(normalized_sequence, verbose=0)[0]
        
        # Decode prediction
        predicted_label, confidence = self.preprocessor.decode_prediction(prediction_proba)
        
        # Track prediction time
        prediction_time = time.time() - start_time
        self.prediction_times.append(prediction_time)
        
        return predicted_label, confidence, prediction_proba
    
    def get_stable_prediction(self):
        """
        Get stable prediction based on recent history.
        
        Returns:
            Tuple of (stable_label, average_confidence) or (None, 0.0)
        """
        if len(self.prediction_history) < 5:
            return None, 0.0
        
        # Get high-confidence predictions
        confident_predictions = [
            (label, conf) for label, conf in self.prediction_history 
            if conf > self.confidence_threshold
        ]
        
        if not confident_predictions:
            return None, 0.0
        
        # Count occurrences of each prediction
        prediction_counts = Counter([pred[0] for pred in confident_predictions])
        
        # Get most common prediction
        most_common_label, count = prediction_counts.most_common(1)[0]
        
        if count >= 3:  # Require at least 3 occurrences
            # Calculate average confidence for this prediction
            confidences = [
                conf for label, conf in confident_predictions 
                if label == most_common_label
            ]
            avg_confidence = sum(confidences) / len(confidences)
            return most_common_label, avg_confidence
        
        return None, 0.0
    
    def draw_interface(self, frame, prediction, confidence, fps, buffer_status):
        """
        Draw user interface on frame.
        
        Args:
            frame: Input video frame
            prediction: Current prediction (can be None)
            confidence: Prediction confidence
            fps: Current frame rate
            buffer_status: Sequence buffer status
            
        Returns:
            Annotated frame with interface
        """
        height, width = frame.shape[:2]
        
        # Create info panel
        panel_height = 150
        panel = np.zeros((panel_height, width, 3), dtype=np.uint8)
        panel[:] = (40, 40, 40)  # Dark gray background
        
        # Main prediction display
        if prediction:
            # Large prediction text
            pred_text = f"Sign: {prediction}"
            conf_text = f"Confidence: {confidence:.2f}"
            
            # Color based on confidence
            if confidence > 0.8:
                color = (0, 255, 0)  # Green
            elif confidence > 0.6:
                color = (0, 255, 255)  # Yellow
            else:
                color = (0, 165, 255)  # Orange
            
            cv2.putText(panel, pred_text, (10, 40), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1.5, color, 3)
            cv2.putText(panel, conf_text, (10, 80), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)
        else:
            cv2.putText(panel, "Show clear ASL sign", (10, 40), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1.2, (150, 150, 150), 2)
            cv2.putText(panel, "Building sequence...", (10, 80), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1.0, (150, 150, 150), 2)
        
        # Performance metrics
        fps_text = f"FPS: {fps:.1f}"
        cv2.putText(panel, fps_text, (width - 200, 40), 
                   cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)
        
        # Buffer status
        buffer_text = f"Buffer: {buffer_status}/{self.sequence_length}"
        cv2.putText(panel, buffer_text, (width - 250, 80), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.8, (200, 200, 200), 2)
        
        # Statistics
        if self.total_frames > 0:
            success_rate = self.successful_predictions / self.total_frames * 100
            stats_text = f"Success: {success_rate:.1f}%"
            cv2.putText(panel, stats_text, (width - 200, 120), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.8, (200, 200, 200), 2)
        
        # Instructions
        instructions = [
            "Controls: SPACE=Reset | R=Clear History | Q=Quit",
            "Show clear ASL alphabet signs to camera"
        ]
        for i, instruction in enumerate(instructions):
            cv2.putText(panel, instruction, (10, 110 + i * 20), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (180, 180, 180), 1)
        
        # Combine frame and panel
        combined_frame = np.vstack([frame, panel])
        
        return combined_frame
    
    def run_recognition(self, save_demo=False):
        """
        Run the real-time ASL recognition system.
        
        Args:
            save_demo: Whether to save demo video
        """
        print("STARTING REAL-TIME ASL RECOGNITION")
        print("="*50)
        print("Controls:")
        print("  SPACE - Reset sequence buffer")
        print("  R - Clear prediction history")
        print("  Q - Quit system")
        print("\nPosition your hand clearly in front of the camera.")
    
        # Initialize video capture
        cap = cv2.VideoCapture(0)
        cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
        cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
        cap.set(cv2.CAP_PROP_FPS, 30)
        
        # Check if camera opened successfully
        if not cap.isOpened():
            print("Error: Could not open webcam")
            print("Please check if your camera is connected and not in use by another application")
            return
        
        # Video writer for demo recording
        out = None
        if save_demo:
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter('asl_demo.mp4', fourcc, 20.0, (640, 630))
        
        try:
            while True:
                frame_start = time.time()
                
                ret, frame = cap.read()
                if not ret:
                    print("Failed to capture frame")
                    break
                
                # Flip frame horizontally for mirror effect
                frame = cv2.flip(frame, 1)
                
                # Initialize prediction variables for this frame
                prediction = None
                confidence = 0.0
                
                # Extract hand landmarks
                landmarks, success, annotated_frame = self.hand_extractor.extract_landmarks(frame)
                
                if success:
                    # Normalize landmarks
                    normalized_landmarks = self.hand_extractor.normalize_landmarks(landmarks)
                    
                    # Add to sequence buffer
                    self.sequence_buffer.append(normalized_landmarks)
                    
                    # Make prediction if buffer is full
                    if len(self.sequence_buffer) == self.sequence_length:
                        try:
                            # Get prediction from current sequence
                            pred_label, pred_conf, _ = self.predict_from_sequence(list(self.sequence_buffer))
                            
                            # Add to prediction history
                            self.prediction_history.append((pred_label, pred_conf))
                            
                            # Get stable prediction
                            stable_pred, stable_conf = self.get_stable_prediction()
                            
                            if stable_pred and stable_conf > self.confidence_threshold:
                                prediction = stable_pred
                                confidence = stable_conf
                                self.successful_predictions += 1
                                
                        except Exception as e:
                            print(f"Prediction error: {e}")
                            # Continue with None prediction
                            pass
                
                # Calculate FPS
                frame_time = time.time() - frame_start
                self.frame_times.append(frame_time)
                avg_frame_time = sum(self.frame_times) / len(self.frame_times)
                fps = 1.0 / avg_frame_time if avg_frame_time > 0 else 0
                
                # Draw interface
                display_frame = self.draw_interface(
                    annotated_frame, prediction, confidence, fps, len(self.sequence_buffer)
                )
                
                # Save frame if recording
                if save_demo and out is not None:
                    out.write(display_frame)
                
                # Display frame
                cv2.imshow('Real-Time ASL Recognition', display_frame)
                
                # Update statistics
                self.total_frames += 1
                
                # Handle keyboard input
                key = cv2.waitKey(1) & 0xFF
                if key == ord('q'):
                    break
                elif key == ord(' '):
                    self.sequence_buffer.clear()
                    print("Sequence buffer reset")
                elif key == ord('r'):
                    self.prediction_history.clear()
                    print("Prediction history cleared")
                
        except KeyboardInterrupt:
            print("\nRecognition stopped by user")
        
        except Exception as e:
            print(f"\nUnexpected error: {e}")
            
        finally:
            # Clean up
            cap.release()
            if save_demo and out is not None:
                out.release()
                print("Demo video saved to asl_demo.mp4")
            cv2.destroyAllWindows()
            
            # Print final statistics
            print(f"\n SESSION STATISTICS:")
            print(f"  - Total frames processed: {self.total_frames:,}")
            print(f"  - Successful predictions: {self.successful_predictions:,}")
            if self.total_frames > 0:
                print(f"  - Success rate: {self.successful_predictions/self.total_frames*100:.1f}%")
            if self.prediction_times:
                print(f"  - Average prediction time: {np.mean(self.prediction_times)*1000:.1f} ms")
            
            print("\nReal-time recognition session completed.")

# Initialize the real-time recognition system
print("INITIALIZING REAL-TIME ASL RECOGNITION SYSTEM")
print("="*60)

realtime_recognizer = RealTimeASLRecognizer(
    model=asl_model,
    preprocessor=preprocessor,
    sequence_length=30,
    confidence_threshold=0.7
)

print("Real-time recognition system initialized!")
print("\nTo start real-time recognition, uncomment and run:")
print("realtime_recognizer.run_recognition(save_demo=True)")

# Uncomment the following line to start real-time recognition
realtime_recognizer.run_recognition(save_demo=True)
````

````python
# Cell 13: Different Method
# Cell 13: Stable Real-Time Recognition System Implementation
"""
Real-time ASL recognition system that balances responsiveness with reliability.
Addresses flickering and instability issues while maintaining good performance.
"""

class StableRealTimeASLRecognizer:
    """
    Real-time ASL recognition system with comprehensive filtering and smoothing.
    - Temporal smoothing and prediction filtering
    - Confidence-based prediction validation
    """
    
    def __init__(self, model, preprocessor, sequence_length=20, confidence_threshold=0.75):
        """
        Initialize the stable real-time recognition system.
        
        Args:
            model: Trained LSTM model
            preprocessor: Data preprocessor with normalization parameters
            sequence_length: Balanced sequence length for stability vs responsiveness
            confidence_threshold: Higher threshold for stability
        """
        self.model = model
        self.preprocessor = preprocessor
        self.sequence_length = sequence_length
        self.confidence_threshold = confidence_threshold
        self.base_confidence_threshold = confidence_threshold
        
        # Initialize MediaPipe hand extractor with balanced settings
        self.hand_extractor = HandLandmarkExtractor(
            static_image_mode=False,  # Video mode
            max_num_hands=1,
            min_detection_confidence=0.7,   # Higher for stability
            min_tracking_confidence=0.6     # Balanced tracking confidence
        )
        
        # Sequence buffer with balanced length
        self.sequence_buffer = deque(maxlen=sequence_length)
        
        # Performance tracking
        self.frame_times = deque(maxlen=20)
        self.prediction_times = deque(maxlen=20)
        
        # Enhanced prediction stability system
        self.prediction_history = deque(maxlen=8)       # Longer history for stability
        self.confidence_history = deque(maxlen=8)       # Track confidence trends
        self.stable_predictions = deque(maxlen=5)       # Recent stable predictions
        
        # Temporal smoothing for display stability
        self.display_prediction = None
        self.display_confidence = 0.0
        self.display_stability_counter = 0
        self.prediction_change_cooldown = 0
        
        # Frame processing control for balanced performance
        self.frame_counter = 0
        self.prediction_counter = 0
        self.landmark_extraction_interval = 1           # Process every frame
        self.prediction_interval = 5                    # Predict every 5th frame
        
        # Adaptive system parameters
        self.consecutive_low_confidence = 0
        self.recent_fps = deque(maxlen=10)
        self.system_load_factor = 1.0
        
        # Quality control
        self.hand_detection_stability = deque(maxlen=10)
        self.landmark_quality_scores = deque(maxlen=10)
        
        # Statistics
        self.total_frames = 0
        self.successful_predictions = 0
        self.stable_display_changes = 0
        
    def calculate_landmark_quality(self, landmarks):
        """
        Calculate quality score for extracted landmarks.
        
        Args:
            landmarks: Extracted hand landmarks
            
        Returns:
            Quality score between 0 and 1
        """
        if landmarks is None or len(landmarks) != 42:
            return 0.0
        
        quality_score = 1.0
        
        # Check for valid numeric values
        if np.any(np.isnan(landmarks)) or np.any(np.isinf(landmarks)):
            return 0.0
        
        # Penalize if values are out of expected range
        if not np.all(landmarks >= -1.5) or not np.all(landmarks <= 1.5):
            quality_score *= 0.7
        
        # Check landmark spread (hand should occupy reasonable space)
        x_coords = landmarks[::2]
        y_coords = landmarks[1::2]
        
        x_range = np.max(x_coords) - np.min(x_coords)
        y_range = np.max(y_coords) - np.min(y_coords)
        
        # Penalize if hand appears too small or too large
        if x_range < 0.1 or y_range < 0.1:
            quality_score *= 0.5
        elif x_range > 0.8 or y_range > 0.8:
            quality_score *= 0.8
        
        # Check for landmark consistency (no extreme jumps)
        landmark_distances = []
        for i in range(0, len(landmarks), 2):
            x, y = landmarks[i], landmarks[i+1]
            landmark_distances.append(np.sqrt(x*x + y*y))
        
        if len(landmark_distances) > 0:
            max_distance = np.max(landmark_distances)
            if max_distance > 1.0:
                quality_score *= 0.6
        
        return min(1.0, max(0.0, quality_score))
    
    def predict_from_sequence_stable(self, landmarks_sequence):
        """
        Generate stable prediction from landmark sequence with quality control.
        
        Args:
            landmarks_sequence: Sequence of hand landmarks
            
        Returns:
            Tuple of (predicted_label, confidence, quality_score)
        """
        start_time = time.time()
        
        try:
            # Validate sequence quality
            if len(landmarks_sequence) < self.sequence_length * 0.8:
                return None, 0.0, 0.0
            
            # Calculate average quality of landmarks in sequence
            quality_scores = [self.calculate_landmark_quality(lm) for lm in landmarks_sequence]
            avg_quality = np.mean(quality_scores)
            
            if avg_quality < 0.3:  # Skip low quality sequences
                return None, 0.0, avg_quality
            
            # Prepare sequence for model input
            sequence_array = np.array([landmarks_sequence])
            
            # Normalize using preprocessor statistics
            normalized_sequence = self.preprocessor._normalize_sequences(sequence_array, fit=False)
            
            # Generate prediction
            prediction_proba = self.model.predict(normalized_sequence, verbose=0)[0]
            
            # Decode prediction
            predicted_label, confidence = self.preprocessor.decode_prediction(prediction_proba)
            
            # Adjust confidence based on sequence quality
            adjusted_confidence = confidence * (0.7 + 0.3 * avg_quality)
            
            # Track prediction time
            prediction_time = time.time() - start_time
            self.prediction_times.append(prediction_time)
            
            return predicted_label, adjusted_confidence, avg_quality
            
        except Exception as e:
            print(f"Prediction error: {e}")
            return None, 0.0, 0.0
    
    def update_stable_prediction(self, new_prediction, new_confidence, quality_score):
        """
        Update stable prediction display with temporal smoothing.
        
        Args:
            new_prediction: New prediction result
            new_confidence: New prediction confidence
            quality_score: Quality score of the prediction
        """
        # Add to prediction history
        if new_prediction is not None:
            self.prediction_history.append(new_prediction)
            self.confidence_history.append(new_confidence)
        
        # Decrease change cooldown
        if self.prediction_change_cooldown > 0:
            self.prediction_change_cooldown -= 1
        
        # Analyze prediction stability
        if len(self.prediction_history) >= 5:
            recent_predictions = list(self.prediction_history)[-5:]
            recent_confidences = list(self.confidence_history)[-5:]
            
            # Count occurrences of each prediction
            prediction_counts = Counter(recent_predictions)
            most_common_pred, occurrence_count = prediction_counts.most_common(1)[0]
            
            # Calculate average confidence for the most common prediction
            relevant_confidences = [
                conf for pred, conf in zip(recent_predictions, recent_confidences)
                if pred == most_common_pred
            ]
            avg_confidence = np.mean(relevant_confidences) if relevant_confidences else 0.0
            
            # Determine if we should update the display
            should_update = False
            
            # High confidence immediate update
            if (new_confidence > 0.9 and 
                quality_score > 0.7 and 
                self.prediction_change_cooldown == 0):
                should_update = True
                self.prediction_change_cooldown = 3  # Prevent rapid changes
            
            # Stable prediction update
            elif (occurrence_count >= 3 and 
                  avg_confidence > self.confidence_threshold and
                  quality_score > 0.5):
                
                # Only update if it's different from current display
                if (self.display_prediction != most_common_pred or
                    abs(self.display_confidence - avg_confidence) > 0.1):
                    should_update = True
                    self.prediction_change_cooldown = 5  # Longer cooldown for stability
            
            # Very stable prediction with lower threshold
            elif (occurrence_count >= 4 and 
                  avg_confidence > self.confidence_threshold * 0.85):
                should_update = True
                self.prediction_change_cooldown = 8
            
            # Apply update if conditions are met
            if should_update:
                self.display_prediction = most_common_pred
                self.display_confidence = avg_confidence
                self.display_stability_counter = 10  # Display for at least 10 frames
                self.stable_display_changes += 1
                
                # Add to stable predictions for tracking
                self.stable_predictions.append((most_common_pred, avg_confidence))
        
        # Decrease display stability counter
        if self.display_stability_counter > 0:
            self.display_stability_counter -= 1
        
        # Clear display if no recent stable predictions
        if (self.display_stability_counter <= 0 and 
            len(self.prediction_history) > 0):
            
            # Check if recent predictions are consistently low confidence
            recent_confs = list(self.confidence_history)[-3:] if len(self.confidence_history) >= 3 else []
            if recent_confs and all(c < self.confidence_threshold * 0.7 for c in recent_confs):
                self.display_prediction = None
                self.display_confidence = 0.0
    
    def adapt_system_parameters(self):
        """
        Dynamically adapt system parameters based on performance.
        """
        # Calculate recent FPS
        if len(self.frame_times) > 0:
            recent_fps = 1.0 / np.mean(self.frame_times)
            self.recent_fps.append(recent_fps)
            
            # Adjust prediction interval based on FPS
            avg_fps = np.mean(self.recent_fps)
            
            if avg_fps < 15:  # Low FPS - reduce load
                self.prediction_interval = min(8, self.prediction_interval + 1)
            elif avg_fps > 25:  # Good FPS - can afford more predictions
                self.prediction_interval = max(4, self.prediction_interval - 1)
        
        # Adjust confidence threshold based on recent success
        if len(self.confidence_history) >= 5:
            recent_avg_conf = np.mean(list(self.confidence_history)[-5:])
            
            if recent_avg_conf < 0.5:  # Very low confidence - lower threshold temporarily
                self.confidence_threshold = max(0.6, self.base_confidence_threshold - 0.1)
            elif recent_avg_conf > 0.8:  # High confidence - can afford higher threshold
                self.confidence_threshold = min(0.85, self.base_confidence_threshold + 0.05)
            else:  # Normal operation
                self.confidence_threshold = self.base_confidence_threshold
    
    def draw_stable_interface(self, frame, fps, buffer_status, quality_score):
        """
        Draw stable user interface with comprehensive information.
        
        Args:
            frame: Input video frame
            fps: Current frame rate
            buffer_status: Sequence buffer status
            quality_score: Current landmark quality score
            
        Returns:
            Annotated frame with stable interface
        """
        height, width = frame.shape[:2]
        
        # Create info panel
        panel_height = 150
        panel = np.zeros((panel_height, width, 3), dtype=np.uint8)
        panel[:] = (40, 40, 45)  # Dark blue-gray background
        
        # Main prediction display with stability indicators
        if self.display_prediction:
            pred_text = f"Sign: {self.display_prediction}"
            conf_text = f"Confidence: {self.display_confidence:.2f}"
            
            # Stability-based color coding
            if self.display_confidence > 0.85:
                color = (0, 255, 100)   # Bright green
                status = "EXCELLENT"
            elif self.display_confidence > 0.75:
                color = (0, 255, 255)   # Yellow
                status = "VERY GOOD"
            elif self.display_confidence > 0.65:
                color = (0, 200, 255)   # Orange
                status = "GOOD"
            else:
                color = (0, 150, 255)   # Light red
                status = "FAIR"
            
            cv2.putText(panel, pred_text, (10, 35), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 3)
            cv2.putText(panel, conf_text, (10, 65), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
            cv2.putText(panel, f"Quality: {status}", (10, 90), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
            
            # Stability indicator
            stability_text = f"Stable: {self.display_stability_counter}"
            cv2.putText(panel, stability_text, (10, 115), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200, 200, 200), 1)
            
        else:
            cv2.putText(panel, "Position hand clearly", (10, 35), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1.0, (180, 180, 180), 2)
            cv2.putText(panel, "Building prediction...", (10, 65), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.8, (150, 150, 150), 2)
            
            # Show buffer status
            buffer_text = f"Buffer: {buffer_status}/{self.sequence_length}"
            cv2.putText(panel, buffer_text, (10, 90), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, (120, 120, 120), 1)
        
        # Performance metrics
        fps_text = f"FPS: {fps:.1f}"
        fps_color = (0, 255, 0) if fps > 20 else (0, 255, 255) if fps > 15 else (0, 100, 255)
        cv2.putText(panel, fps_text, (width - 150, 35), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.8, fps_color, 2)
        
        # Quality indicator
        quality_text = f"Quality: {quality_score:.2f}"
        quality_color = (0, 255, 0) if quality_score > 0.7 else (0, 255, 255) if quality_score > 0.5 else (0, 100, 255)
        cv2.putText(panel, quality_text, (width - 180, 65), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, quality_color, 2)
        
        # System status
        if len(self.prediction_times) > 0:
            avg_pred_time = np.mean(self.prediction_times) * 1000
            pred_text = f"Pred: {avg_pred_time:.0f}ms"
            cv2.putText(panel, pred_text, (width - 150, 95), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200, 200, 200), 1)
        
        # Success rate
        if self.total_frames > 0:
            success_rate = self.stable_display_changes / (self.total_frames / 30) if self.total_frames > 30 else 0
            success_text = f"Stable: {success_rate:.1f}/min"
            cv2.putText(panel, success_text, (width - 180, 125), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (180, 180, 180), 1)
        
        # Instructions
        instructions = [
            "Controls: SPACE=Reset | R=Clear | Q=Quit",
            "Hold sign steady for best recognition"
        ]
        for i, instruction in enumerate(instructions):
            cv2.putText(panel, instruction, (10, 135 + i * 10), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.45, (160, 160, 160), 1)
        
        # Combine frame and panel
        combined_frame = np.vstack([frame, panel])
        
        return combined_frame
    
    def run_stable_recognition(self, save_demo=False):
        """
        Run the stable real-time ASL recognition system.
        
        Args:
            save_demo: Whether to save demo video
        """
        print("STARTING STABLE REAL-TIME ASL RECOGNITION")
        print("="*50)
        print("Stability Features:")
        print(f"  - Balanced sequence length: {self.sequence_length} frames")
        print(f"  - Temporal smoothing with prediction filtering")
        print(f"  - Quality-based confidence adjustment")
        print(f"  - Adaptive threshold management: {self.confidence_threshold}")
        print(f"  - Prediction change cooldown system")
        print()
        print("Controls:")
        print("  SPACE - Reset sequence buffer")
        print("  R - Clear prediction history")
        print("  Q - Quit system")
        print("\nHold signs steady for best recognition.")
    
        # Initialize video capture
        cap = cv2.VideoCapture(0)
        cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
        cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
        cap.set(cv2.CAP_PROP_FPS, 30)
        
        # Check if camera opened successfully
        if not cap.isOpened():
            print("Error: Could not open webcam")
            print("Please check if your camera is connected and not in use by another application")
            return
        
        # Video writer for demo recording
        out = None
        if save_demo:
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter('asl_demo_stable.mp4', fourcc, 20.0, (640, 630))
        
        try:
            while True:
                frame_start = time.time()
                
                ret, frame = cap.read()
                if not ret:
                    print("Failed to capture frame")
                    break
                
                # Flip frame horizontally for mirror effect
                frame = cv2.flip(frame, 1)
                
                # Initialize frame variables
                annotated_frame = frame.copy()
                current_quality = 0.0
                
                self.frame_counter += 1
                
                # Process landmarks every frame for smooth visual feedback
                if self.frame_counter % self.landmark_extraction_interval == 0:
                    landmarks, success, annotated_frame = self.hand_extractor.extract_landmarks(frame)
                    
                    if success:
                        # Calculate landmark quality
                        current_quality = self.calculate_landmark_quality(landmarks)
                        self.landmark_quality_scores.append(current_quality)
                        
                        # Add to detection stability tracking
                        self.hand_detection_stability.append(1)
                        
                        if current_quality > 0.4:  # Only use good quality landmarks
                            # Normalize landmarks
                            normalized_landmarks = self.hand_extractor.normalize_landmarks(landmarks)
                            
                            # Add to sequence buffer
                            self.sequence_buffer.append(normalized_landmarks)
                        
                    else:
                        self.hand_detection_stability.append(0)
                        current_quality = 0.0
                
                # Make predictions at controlled intervals
                self.prediction_counter += 1
                if (self.prediction_counter % self.prediction_interval == 0 and 
                    len(self.sequence_buffer) >= self.sequence_length * 0.8):
                    
                    try:
                        # Get prediction from current sequence
                        pred_label, pred_conf, quality = self.predict_from_sequence_stable(
                            list(self.sequence_buffer)
                        )
                        
                        # Update stable prediction system
                        self.update_stable_prediction(pred_label, pred_conf, quality)
                        
                        if pred_label and pred_conf > self.confidence_threshold:
                            self.successful_predictions += 1
                            
                    except Exception as e:
                        print(f"Prediction error: {e}")
                
                # Adapt system parameters periodically
                if self.frame_counter % 30 == 0:  # Every 30 frames
                    self.adapt_system_parameters()
                
                # Calculate FPS
                frame_time = time.time() - frame_start
                self.frame_times.append(frame_time)
                avg_frame_time = sum(self.frame_times) / len(self.frame_times)
                fps = 1.0 / avg_frame_time if avg_frame_time > 0 else 0
                
                # Use average quality from recent measurements
                avg_quality = np.mean(self.landmark_quality_scores) if self.landmark_quality_scores else 0.0
                
                # Draw stable interface
                display_frame = self.draw_stable_interface(
                    annotated_frame, fps, len(self.sequence_buffer), avg_quality
                )
                
                # Save frame if recording
                if save_demo and out is not None:
                    out.write(display_frame)
                
                # Display frame
                cv2.imshow('Stable Real-Time ASL Recognition', display_frame)
                
                # Update statistics
                self.total_frames += 1
                
                # Handle keyboard input
                key = cv2.waitKey(1) & 0xFF
                if key == ord('q'):
                    break
                elif key == ord(' '):
                    self.sequence_buffer.clear()
                    self.prediction_history.clear()
                    self.confidence_history.clear()
                    self.display_prediction = None
                    self.display_confidence = 0.0
                    print("System reset")
                elif key == ord('r'):
                    self.prediction_history.clear()
                    self.confidence_history.clear()
                    self.stable_predictions.clear()
                    self.display_prediction = None
                    self.display_confidence = 0.0
                    print("Prediction history cleared")
                
        except KeyboardInterrupt:
            print("\nRecognition stopped by user")
        
        except Exception as e:
            print(f"\nUnexpected error: {e}")
            
        finally:
            # Clean up
            cap.release()
            if save_demo and out is not None:
                out.release()
                print("Demo video saved to asl_demo_stable.mp4")
            cv2.destroyAllWindows()
            
            # Print comprehensive statistics
            print(f"\n STABLE SESSION STATISTICS:")
            print(f"  - Total frames processed: {self.total_frames:,}")
            print(f"  - Successful predictions: {self.successful_predictions:,}")
            print(f"  - Stable display changes: {self.stable_display_changes:,}")
            
            if self.total_frames > 0:
                prediction_rate = self.successful_predictions / (self.total_frames / 30)  # per 30 frames
                stability_rate = self.stable_display_changes / (self.total_frames / 30)   # per 30 frames
                print(f"  - Prediction rate: {prediction_rate:.1f} per 30 frames")
                print(f"  - Stability rate: {stability_rate:.1f} changes per 30 frames")
                
            if self.prediction_times:
                print(f"  - Average prediction time: {np.mean(self.prediction_times)*1000:.1f} ms")
                
            if self.frame_times:
                print(f"  - Average FPS: {1/np.mean(self.frame_times):.1f}")
                
            if self.landmark_quality_scores:
                print(f"  - Average landmark quality: {np.mean(self.landmark_quality_scores):.2f}")
            
            print("\nStable real-time recognition session completed.")

# Initialize the real-time recognition system
print("INITIALIZING STABLE REAL-TIME ASL RECOGNITION SYSTEM")
print("="*60)

stable_recognizer = StableRealTimeASLRecognizer(
    model=asl_model,
    preprocessor=preprocessor,
    sequence_length=20,      # Balanced length for stability vs responsiveness
    confidence_threshold=0.75  # Higher threshold for stability
)

print("Stable real-time recognition system initialized!")
print("\nStability features:")
print("  - Temporal smoothing with prediction filtering")
print("  - Quality-based confidence adjustment")
print("  - Adaptive system parameter management") 
print("  - Prediction change cooldown system")
print("  - Comprehensive error handling and recovery")
print("\nTo start real-time recognition, run:")
print("stable_recognizer.run_stable_recognition(save_demo=True)")

# To start real-time recognition
stable_recognizer.run_stable_recognition(save_demo=True)
````

## Part D: Testing Methodologies

### Appropriate Testing Methods

**Quantitative Evaluation Metrics:** Standard metrics include accuracy, precision, recall, and F1-score for isolated signs, and Word Error Rate (WER) and BLEU scores for continuous sign recognition and translation. Frame-level accuracy assesses sign identification within sequences, while sequence-level accuracy evaluates sentence recognition (Camgoz et al., 2020).

**Cross-Validation Strategies:** K-fold cross-validation ensures robust performance across data subsets, while leave-one-signer-out validation tests generalization to new users. Temporal cross-validation simulates real-world scenarios by splitting data chronologically (Adeyanju et al., 2021).

**Robustness Testing:** Evaluates performance across lighting conditions, backgrounds, camera angles, user demographics, signing speeds, and occlusions. This ensures systems are practical in diverse environments (Rahman et al., 2024).

**Real-Time Performance Evaluation:** Measures end-to-end latency (<100ms for real-time, <50ms for critical applications) and frame rate consistency (>20fps). Memory usage profiling identifies bottlenecks for deployment (Lugaresi et al., 2019).

### Expected vs. Actual Results

Based on literature, expected benchmarks include:

- Isolated sign recognition: >85% accuracy.
- Continuous sign recognition: 70–75% accuracy.
- Latency: <100ms for real-time applications.
- Frame rate: >20fps on standard hardware.

Actual results from studies align closely, with MediaPipe + LSTM achieving 82–89% for isolated signs and 78–88% for continuous signs, though performance gaps occur with complex sentences or environmental variations (Sahoo et al., 2022; Srivastava et al., 2024).

### Performance Analysis

Disparities often stem from confusion between visually similar signs (e.g., ASL letters 'M', 'N', 'T') or challenges at sign boundaries in continuous recognition. Robustness testing reveals performance degradation under low-light conditions (15% accuracy loss) or cluttered backgrounds (8% loss), highlighting deployment constraints (Ilham et al., 2023).

## Part E: Results Evaluation

### Accuracy Assessment and Confidence Analysis

Accuracy interpretation depends on application context:

**Isolated Signs:** Accuracies above 95% (e.g., 99.98% for ASL alphabet) support educational and vocabulary learning applications (Sahoo et al., 2022).

**Continuous Signs:** Accuracies around 80–85% (e.g., 88.23% with MediaPipe + LSTM) indicate readiness for assistive tools but limitations for critical communication (Srivastava et al., 2024).

Cross-validation shows stable performance (standard deviation ~2%), but leave-one-signer-out testing reveals 5% accuracy degradation, suggesting the need for diverse training data or personalization (Adeyanju et al., 2021).

### Test Case Analysis

Confusion matrices indicate high accuracy for simple sign combinations but reduced performance for complex grammatical structures, particularly at sign boundaries. Real-time metrics (e.g., 45ms latency, 25fps) confirm suitability for interactive applications, though memory usage scales with sequence length, requiring optimization for extended use (Lugaresi et al., 2019).

### Problem-Solving Effectiveness

The systems effectively support educational and assistive applications but fall short for comprehensive communication replacement due to limitations in contextual understanding and non-manual feature integration. High-confidence predictions (>0.85) achieve 96% accuracy, enabling reliable feedback mechanisms (Ilham et al., 2023).

### Limitations and Future Directions

**Data-Related Limitations:** Dataset biases from limited signer diversity and environmental variations.

**Model-Related Limitations:** Overfitting risks, computational complexity, and limited interpretability in 3D CNNs.

**Future Enhancements:** Expanding multimodal datasets, optimizing transformers for edge devices, and incorporating user adaptation mechanisms.

### Social Impact and Ethical Considerations

Sign language recognition enhances accessibility, fostering independence for deaf individuals. However, deployment must address privacy concerns with facial data and manage user expectations to avoid over-reliance in critical scenarios (Rahman et al., 2024).

````python
def convert_model_via_command_line(model_path, output_dir):
    """
    Convert model using command-line tool instead of Python API
    """
    print("\nAPPROACH 2: Command-Line Model Conversion")
    print("=" * 50)
    
    try:
        # Ensure we have the command-line tool
        print("Step 1: Checking tensorflowjs_converter...")
        result = subprocess.run([sys.executable, "-m", "tensorflowjs.converters.converter", "--help"], 
                              capture_output=True, text=True)
        
        if result.returncode != 0:
            print("  Installing tensorflowjs command-line tools...")
            subprocess.check_call([sys.executable, "-m", "pip", "install", "tensorflowjs[wizard]"])
        
        print("  ✓ Command-line converter available")
        
        # Convert the model
        print(f"\nStep 2: Converting model from {model_path} to {output_dir}...")
        
        conversion_command = [
            sys.executable, "-m", "tensorflowjs.converters.converter",
            "--input_format", "keras",
            "--output_format", "tfjs_graph_model",
            model_path,
            output_dir
        ]
        
        result = subprocess.run(conversion_command, capture_output=True, text=True)
        
        if result.returncode == 0:
            print("  ✓ Model converted successfully via command line!")
            return True
        else:
            print(f"  ❌ Command-line conversion failed: {result.stderr}")
            return False
            
    except Exception as e:
        print(f"  ❌ Command-line approach failed: {e}")
        return False
````
